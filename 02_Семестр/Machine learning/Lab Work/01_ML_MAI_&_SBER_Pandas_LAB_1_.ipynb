   - При необходимости, изменить тип данных колонок.



````python
from typing import Dict

def update_data_types(df: pd.DataFrame) -> pd.DataFrame:
    """
    Description:
        Обновляет типы данных в DataFrame в соответствии с ожидаемыми типами.

    Args:
        df: Исходный DataFrame с данными.

    Returns:
        DataFrame с обновленными типами данных.

    Raises:
        ValueError: Если тип данных не может быть преобразован.
    """
    # Проверим текущие типы данных
    data_types = df.dtypes

    # Ожидаемые типы данных
    expected_types: Dict[str, str] = {
        "CRIM":    "float64",
        "ZN":      "float64",
        "INDUS":   "float64",
        "CHAS":    "category",  # Бинарная категориальная переменная
        "NOX":     "float64",
        "RM":      "float64",
        "AGE":     "float64",
        "DIS":     "float64",
        "RAD":     "int64",     # Целочисленный индекс
        "TAX":     "float64",
        "PTRATIO": "float64",
        "B":       "float64",
        "LSTAT":   "float64",
        "TARGET":  "float64"    # Медианная стоимость домов
    }

    # Изменим тип данных, если он не соответствует ожидаемому
    for col, expected_type in expected_types.items():
        if str(df[col].dtype) != expected_type:
            try:
                if expected_type == "category":
                    df[col] = df[col].astype("category")
                else:
                    df[col] = df[col].astype(expected_type)
            except ValueError as e:
                raise ValueError(f"Не удалось преобразовать столбец {col} в тип {expected_type}: {e}")

    # Проверим обновленные типы данных
    updated_data_types = df.dtypes

    return df
````

4. **Обработка выбросов**: [+1 балл]
   - Определить пороги для выбросов на основе статистических мер (например, IQR)
   - Решить, удалять ли выбросы или корректировать их значения.

````python
def detect_outliers_iqr(data: pd.DataFrame, columns: list) -> pd.DataFrame:
    """
    Description:
        Определяет выбросы в данных на основе метода межквартильного размаха (IQR).

    Args:
        data: Исходный DataFrame
        columns: Список числовых колонок для анализа выбросов

    Returns:
        DataFrame с флагами выбросов (1 - выброс, 0 - нет выброса)

    Examples:
        >>> detect_outliers_iqr(df, ["CRIM", "ZN", "INDUS"])
    """
    outlier_flags = pd.DataFrame(index=data.index)

    for col in columns:
        Q1 = data[col].quantile(0.25)  # Первый квартиль
        Q3 = data[col].quantile(0.75)  # Третий квартиль
        IQR = Q3 - Q1                  # Межквартильный размах

        lower_bound = Q1 - 1.5 * IQR   # Нижний порог
        upper_bound = Q3 + 1.5 * IQR   # Верхний порог

        # Флаг выбросов (1 - выброс, 0 - нет выброса)
        outlier_flags[col] = ((data[col] < lower_bound) | (data[col] > upper_bound)).astype(int)

    return outlier_flags

# Определяем выбросы для числовых столбцов
outliers_df = detect_outliers_iqr(df, numeric_columns)

# Подсчитываем количество выбросов по каждому признаку
outlier_counts = outliers_df.sum()


outlier_counts
````

### Выбранный подход: Winsorization + логарифмическое преобразование  

Мы реализуем обработку выбросов с использованием двух подходов:
- **Для переменных с сильными выбросами (`CRIM`, `ZN`, `B`, `TARGET`)** – логарифмическое преобразование.
- **Для умеренных выбросов (`PTRATIO`, `LSTAT`, `DIS`, `RM`)** – обрезка значений (Winsorization).

````python
import numpy as np
from scipy.stats.mstats import winsorize

# Функция для обработки выбросов
def handle_outliers(data: pd.DataFrame, columns: list) -> pd.DataFrame:
    """
    Description:
        Обрабатывает выбросы в числовых столбцах с помощью логарифмического преобразования
        и Winsorization (обрезка выбросов).

    Args:
        data: Исходный DataFrame
        columns: Список числовых колонок для обработки

    Returns:
        DataFrame с обработанными выбросами

    Examples:
        >>> handle_outliers(df, ["CRIM", "ZN", "B", "TARGET"])
    """
    data = data.copy()

    # Логарифмическое преобразование (для сильно асимметричных признаков)
    log_transform_cols = ["CRIM", "ZN", "B", "TARGET"]
    for col in log_transform_cols:
        data[col] = np.log1p(data[col])  # log(1 + x) во избежание log(0)

    # Winsorization (для умеренных выбросов)
    winsorize_cols = ["PTRATIO", "LSTAT", "DIS", "RM"]
    for col in winsorize_cols:
        data[col] = winsorize(data[col], limits=[0.05, 0.05])  # 5% обрезка

    return data

# Применяем обработку выбросов
df_cleaned = handle_outliers(df, numeric_columns)
````

### Что такое IQR и как его использовать для обработки выбросов:

**IQR (Interquartile Range)**, или межквартильный размах, это мера статистического разброса и используется для определения выбросов. IQR рассчитывается как разность между 75-м и 25-м процентилями данных (Q3 и Q1 соответственно).

Выбросы часто определяются как значения, которые находятся за пределами следующих границ:
- Нижняя граница: Q1 - 1.5 * IQR
- Верхняя граница: Q3 + 1.5 * IQR

#### Примерный код на Python с использованием pandas:

```python
import pandas as pd

# Загрузка данных
df = pd.read_csv('path_to_file.csv')

# Рассчитываем IQR для колонки 'example_column'
Q1 = df['example_column'].quantile(0.25)
Q3 = df['example_column'].quantile(0.75)
IQR = Q3 - Q1

# Определяем границы для выбросов
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Фильтрация данных для удаления или корректировки выбросов
filtered_df = df[(df['example_column'] >= lower_bound) & (df['example_column'] <= upper_bound)]
```

````python
# Статистический анализ
from scipy.stats import pearsonr

# Функция для построения диаграммы рассеяния
def plot_scatter_rm_vs_target(data: pd.DataFrame) -> None:
    """
    Description:
        Строит диаграмму рассеяния для анализа зависимости между средним
        количеством комнат в доме (RM) и медианной стоимостью жилья (TARGET).

    Args:
        data: DataFrame, содержащий переменные RM и TARGET.

    Returns:
        None

    Raises:
        KeyError: Если в DataFrame отсутствуют требуемые столбцы.

    Examples:
        >>> plot_scatter_rm_vs_target(df)
    """
    if "RM" not in data.columns or "TARGET" not in data.columns:
        raise KeyError("DataFrame должен содержать столбцы 'RM' и 'TARGET'.")

    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=data["RM"], y=data["TARGET"])
    plt.xlabel("Среднее количество комнат (RM)")
    plt.ylabel("Логарифм медианной стоимости жилья (TARGET)")
    plt.title("Диаграмма рассеяния RM vs. TARGET")
    plt.show()


# Функция для расчета коэффициента корреляции Пирсона
def calculate_pearson_correlation(data: pd.DataFrame) -> pd.DataFrame:
    """
    Description:
        Вычисляет коэффициент корреляции Пирсона между средним числом комнат
        (RM) и медианной стоимостью жилья (TARGET).

    Args:
        data: DataFrame, содержащий переменные RM и TARGET.

    Returns:
        DataFrame с коэффициентом корреляции и p-значением.

    Raises:
        KeyError: Если в DataFrame отсутствуют требуемые столбцы.

    Examples:
        >>> calculate_pearson_correlation(df)
    """
    if "RM" not in data.columns or "TARGET" not in data.columns:
        raise KeyError("DataFrame должен содержать столбцы 'RM' и 'TARGET'.")

    correlation, p_value = pearsonr(data["RM"], data["TARGET"])

    return pd.DataFrame(
        {"Метрика": ["Коэффициент Пирсона", "p-значение"], "Значение": [correlation, p_value]}
    )


# Вызов функций для анализа
plot_scatter_rm_vs_target(df_cleaned)
correlation_result = calculate_pearson_correlation(df_cleaned)

# Отображение результатов корреляции
print("Корреляция RM и TARGET", correlation_result)
````

````python
def calculate_correlation_matrix(dataframe: pd.DataFrame, columns: list[str]) -> pd.DataFrame:
    """
    Description:
        Вычисляет коэффициенты корреляции Пирсона для указанных столбцов в DataFrame.

    Args:
        dataframe: DataFrame, содержащий данные.
        columns: Список столбцов для вычисления корреляции.

    Returns:
        DataFrame с коэффициентами корреляции.

    Raises:
        KeyError: Если указанные столбцы отсутствуют в DataFrame.

    Examples:
        >>> df = pd.DataFrame({'tax': [1, 2, 3], 'ptratio': [4, 5, 6]})
        >>> calculate_correlation_matrix(df, ['tax', 'ptratio'])
           tax  ptratio
    tax     1.0      1.0
    ptratio 1.0      1.0
    """
    return dataframe[columns].corr()

def plot_heatmap(corr_matrix: pd.DataFrame, title: str) -> None:
    """
    Description:
        Визуализирует heatmap для матрицы корреляции.

    Args:
        corr_matrix: DataFrame с коэффициентами корреляции.
        title: Заголовок графика.

    Returns:
        None

    Raises:
        ValueError: Если corr_matrix не является DataFrame.

    Examples:
        >>> corr_matrix = pd.DataFrame({'tax': [1.0, 1.0], 'ptratio': [1.0, 1.0]}, index=['tax', 'ptratio'])
        >>> plot_heatmap(corr_matrix, 'Корреляционная матрица TAX и PTRATIO')
    """
    plt.figure(figsize=(6, 4))
    sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".3f", linewidths=0.5)
    plt.title(title)
    plt.show()

# Вычисление коэффициента корреляции Пирсона
corr_matrix: pd.DataFrame = calculate_correlation_matrix(df_cleaned, ['TAX', 'PTRATIO'])

# Визуализация heatmap
plot_heatmap(corr_matrix, 'Корреляционная матрица TAX и PTRATIO')

# Вывод коэффициента корреляции
print(corr_matrix)
````

Корреляция показывает степень связи между двумя переменными, но не доказывает, что одна переменная вызывает изменения в другой. Это статистическая мера, указывающая на направление и силу взаимосвязи, но она не объясняет механизм взаимодействия.

2. **Причинно-следственная связь** указывает, что одно явление вызывает изменения в другом. Например, если рост процента населения с низким социальным статусом приводит к снижению стоимости жилья из-за ухудшения социальной и экономической привлекательности района, то это причинно-следственная связь.
3. **Корреляция ≠ Причинность**: высокая корреляция не означает, что одно значение вызывает другое. Влияние могут оказывать скрытые переменные, например, экономическое развитие или уровень преступности.
