## –ó–∞–¥–∞—á–∞ –£—Ä–∞–≤–Ω–µ–Ω–∏–µ ([—Å—Å—ã–ª–∫–∞](https://codeforces.com/edu/course/2/lesson/6/2/practice/contest/283932/problem/E)) [1.5 –±–∞–ª–ª–∞]


–ù–∞–π–¥–∏—Ç–µ —Ç–∞–∫–æ–µ —á–∏—Å–ª–æ $x$, —á—Ç–æ $x^2 + \sqrt{x} = c$

### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
–í –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ —Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ $C$ $(1.0 \leq C \leq 10^{10})$.

### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
–í—ã–≤–µ–¥–∏—Ç–µ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äî –∏—Å–∫–æ–º—ã–π $x$. –û—Ç–≤–µ—Ç –±—É–¥–µ—Ç –ø—Ä–∏–∑–Ω–∞–Ω –≤–µ—Ä–Ω—ã–º, –µ—Å–ª–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –∏–ª–∏ –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å –Ω–µ –±–æ–ª–µ–µ $10^{-6}$.

### –ü—Ä–∏–º–µ—Ä—ã  

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  

2.0

#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  

1.0

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  

15.6

#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  

3.698232168829691



````python
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –≤—Ä–µ–º–µ–Ω–∏ - –û(TODO)
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –ø–∞–º—è—Ç–∏ - O(TODO)

# –í–ê–® –ö–û–î
import sys
import math
from typing import List

def solve() -> None:
    """
    Description:
        –†–µ—à–∞–µ—Ç —É—Ä–∞–≤–Ω–µ–Ω–∏–µ mid * mid + sqrt(mid) = c —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –º–µ—Ç–æ–¥–∞ –±–∏–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞.

    Args:
        None

    Returns:
        None

    Raises:
        None

    Examples:
        >>> solve()
        –í–≤–æ–¥: 2
        –í—ã–≤–æ–¥: 1.0
    """
    input_data: List[str] = sys.stdin.read().strip().split()
    if not input_data:
        return

    c: float = float(input_data[0])

    lo: float = 0.0
    hi: float = c

    for _ in range(100):
        mid: float = (lo + hi) / 2
        if mid * mid + math.sqrt(mid) > c:
            hi = mid
        else:
            lo = mid

    print(lo)
````

### –í–ê–®–ê –∏–Ω—Ç—É–∏—Ü–∏—è

–ù–∞—à–∞ –∑–∞–¥–∞—á–∞ —Å–æ—Å—Ç–æ–∏—Ç –≤ –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–∏ —á–∏—Å–ª–∞ $x$, —Ç–∞–∫–æ–≥–æ —á—Ç–æ $x^2 + \sqrt{x} = c$, –≥–¥–µ $c$ ‚Äî –∑–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ. –§—É–Ω–∫—Ü–∏—è $f(x) = x^2 + \sqrt{x}$ —Å—Ç—Ä–æ–≥–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç –¥–ª—è $x \ge 0$, —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –ø—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ $x$ –∑–Ω–∞—á–µ–Ω–∏–µ $f(x)$ —Ç–∞–∫–∂–µ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è, –∏, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –∏–º–µ–µ—Ç –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ –≤ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º–æ–º –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ.

–†–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ —Å–≤–æ–¥–∏—Ç—Å—è –∫ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—é –º–µ—Ç–æ–¥–∞ –¥–∏—Ö–æ—Ç–æ–º–∏–∏, –∏–ª–∏ –¥–≤–æ–∏—á–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞. –ù–∞—á–∏–Ω–∞–µ–º —Å –≤—ã–±–æ—Ä–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ $[0, c]$, –ø–æ—Å–∫–æ–ª—å–∫—É –ø—Ä–∏ $x = 0$ —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞–≤–Ω–∞ $0$, –∞ –ø—Ä–∏ $x = c$ –∑–Ω–∞—á–µ–Ω–∏–µ $f(x)$ –≥–∞—Ä–∞–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å—à–µ $c$ (–ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ $c \ge 1$). –ù–∞ –∫–∞–∂–¥–æ–º —à–∞–≥–µ –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è —Å—Ä–µ–¥–Ω—è—è —Ç–æ—á–∫–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞, –∏ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ç–æ–≥–æ, –±–æ–ª—å—à–µ –∏–ª–∏ –º–µ–Ω—å—à–µ –≤—ã—á–∏—Å–ª–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ $f(x)$ —á–∏—Å–ª–∞ $c$, –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç —Å—É–∂–µ–Ω–∏–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞, —á—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –ø–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ –ø—Ä–∏–±–ª–∏–∑–∏—Ç—å—Å—è –∫ –∏—Å–∫–æ–º–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é —Å –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–π —Ç–æ—á–Ω–æ—Å—Ç—å—é.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –¥–∞–Ω–Ω–∞—è –∑–∞–¥–∞—á–∞ —Å–≤–æ–¥–∏—Ç—Å—è –∫ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –º–µ—Ç–æ–¥–∞ –¥–∏—Ö–æ—Ç–æ–º–∏–∏, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∑–≤–æ–ª—è–µ—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ –∏ —Ç–æ—á–Ω–æ –Ω–∞—Ö–æ–¥–∏—Ç—å —Ä–µ—à–µ–Ω–∏–µ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –∑–∞ —Å—á–µ—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ–≥–æ –¥–µ–ª–µ–Ω–∏—è –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –ø–æ–∏—Å–∫–∞. –¢–∞–∫–æ–π –ø–æ–¥—Ö–æ–¥ –æ–±–µ—Å–ø–µ—á–∏–≤–∞–µ—Ç –±—ã—Å—Ç—Ä–æ–µ —Å—É–∂–µ–Ω–∏–µ –æ–±–ª–∞—Å—Ç–∏ –ø–æ–∏—Å–∫–∞ –¥–æ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –Ω—É–∂–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è $x$, —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–µ–≥–æ —É—Å–ª–æ–≤–∏—é $x^2 + \sqrt{x} = c$.

### –í–ê–® —Å–∫—Ä–∏–Ω –û–ö–Ω—É—Ç–æ–π –∑–∞–¥–∞—á–∏

## –ó–∞–¥–∞—á–∞ –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –≤—ã—á–µ—Ä–∫–∏–≤–∞–Ω–∏–µ ([—Å—Å—ã–ª–∫–∞](https://codeforces.com/edu/course/2/lesson/6/2/practice/contest/283932/problem/F)) [1.5 –±–∞–ª–ª–∞]

–ü–µ—Ç—è —Ö–æ—á–µ—Ç –ø–æ–ª—É—á–∏—Ç—å —Å–ª–æ–≤–æ $p$ –∏–∑ —Å–ª–æ–≤–∞ $t$, –≤—ã—á–µ—Ä–∫–∏–≤–∞—è –±—É–∫–≤—ã –≤ –∑–∞–¥–∞–Ω–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ. –ü–æ—Å–ª–µ –∫–∞–∂–¥–æ–≥–æ —É–¥–∞–ª–µ–Ω–∏—è –Ω—É–º–µ—Ä–∞—Ü–∏—è –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –±—É–∫–≤ –Ω–µ –º–µ–Ω—è–µ—Ç—Å—è.

–í–∞—Å—è, –µ–≥–æ –±—Ä–∞—Ç, —Ö–æ—á–µ—Ç –æ—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –ü–µ—Ç—Ä–∞ –∫–∞–∫ –º–æ–∂–Ω–æ –ø–æ–∑–∂–µ, –Ω–æ –¥–æ —Ç–æ–≥–æ –º–æ–º–µ–Ω—Ç–∞, –∫–æ–≥–¥–∞ —Å–ª–æ–≤–æ $p$ —Å—Ç–∞–Ω–µ—Ç –Ω–µ–¥–æ—Å—Ç–∏–∂–∏–º—ã–º. –í–∞—à–∞ –∑–∞–¥–∞—á–∞ ‚Äî –Ω–∞–π—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±—É–∫–≤, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–µ—Ç —É–¥–∞–ª–∏—Ç—å –ü–µ—Ç—è, –ø—Ä–µ–∂–¥–µ —á–µ–º –í–∞—Å—è –µ–≥–æ –æ—Å—Ç–∞–Ω–æ–≤–∏—Ç.

### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
- –ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Å—Ç—Ä–æ–∫—É $t$ $(1 \leq |p| < |t| \leq 200000)$.
- –í—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Å—Ç—Ä–æ–∫—É $p$.
- –¢—Ä–µ—Ç—å—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫—É $a_1, a_2, \dots, a_{|t|}$ ‚Äî –ø–æ—Ä—è–¥–æ–∫, –≤ –∫–æ—Ç–æ—Ä–æ–º –ü–µ—Ç—è –≤—ã—á–µ—Ä–∫–∏–≤–∞–µ—Ç –±—É–∫–≤—ã ($1 \leq a_i \leq |t|$, –≤—Å–µ $a_i$ —Ä–∞–∑–ª–∏—á–Ω—ã).

### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
–í—ã–≤–µ–¥–∏—Ç–µ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äî –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –±—É–∫–≤, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–µ—Ç –≤—ã—á–µ—Ä–∫–Ω—É—Ç—å –ü–µ—Ç—è.

### –ü—Ä–∏–º–µ—Ä—ã  

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
ababcba
abb
5 3 4 1 7 6 2
```
#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
3
```

````python
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –≤—Ä–µ–º–µ–Ω–∏ - –û(TODO)
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –ø–∞–º—è—Ç–∏ - O(TODO)

# –í–ê–® –ö–û–î
import sys
from typing import List

def solve() -> None:
    """
    Description:
        –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏.
        –ß–∏—Ç–∞–µ—Ç –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ, –ø—Ä–æ–≤–µ—Ä—è–µ—Ç –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–µ–Ω–∏—è —Å—Ç—Ä–æ–∫–∏ p –∏–∑ —Å—Ç—Ä–æ–∫–∏ t
        –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –ø–µ—Ä–≤—ã—Ö k —Å–∏–º–≤–æ–ª–æ–≤ –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É.

    Args:
        None

    Returns:
        None

    Raises:
        None

    Examples:
        >>> solve()
        # –í–≤–æ–¥: 3 ab 1 2 3
        # –í—ã–≤–æ–¥: 1
    """
    input_data = sys.stdin.read().split()
    if not input_data:
        return

    t = input_data[0]
    p = input_data[1]
    n = len(t)
    a = list(map(int, input_data[2:]))

    def possible(k: int) -> bool:
        """
        Description:
            –ü—Ä–æ–≤–µ—Ä—è–µ—Ç, –º–æ–∂–Ω–æ –ª–∏ –ø–æ–ª—É—á–∏—Ç—å —Å–ª–æ–≤–æ p –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –ø–µ—Ä–≤—ã—Ö k —Å–∏–º–≤–æ–ª–æ–≤
            –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É.

        Args:
            k: –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–¥–∞–ª—è–µ–º—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤

        Returns:
            True, –µ—Å–ª–∏ –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Å–ª–æ–≤–æ p, –∏–Ω–∞—á–µ False

        Raises:
            None

        Examples:
            >>> possible(1)
            True
        """
        # –ò—Å–ø–æ–ª—å–∑—É–µ–º –∏–Ω–¥–µ–∫—Å–∞—Ü–∏—é —Å 1 –¥–æ n
        removed = [False] * (n + 1)
        for i in range(k):
            removed[a[i]] = True

        j = 0               # —É–∫–∞–∑–∞—Ç–µ–ª—å –Ω–∞ —Å–∏–º–≤–æ–ª—ã —Å—Ç—Ä–æ–∫–∏ p
        # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å—Ç—Ä–æ–∫–µ t, –ø—Ä–æ–ø—É—Å–∫–∞—è —É–¥–∞–ª—ë–Ω–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã
        for i in range(1, n + 1):
            if not removed[i]:
                if j < len(p) and t[i - 1] == p[j]:
                    j += 1
                    if j == len(p):
                        break

        return j == len(p)

    low, high = 0, n
    ans = 0
    # –ë–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫ (–º–µ—Ç–æ–¥ –¥–∏—Ö–æ—Ç–æ–º–∏–∏) –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —É–¥–∞–ª—è–µ–º—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤
    while low <= high:
        mid = (low + high) // 2
        if possible(mid):
            ans = mid
            low = mid + 1
        else:
            high = mid - 1

    sys.stdout.write(str(ans))
````

–°—É—Ç—å —Ä–µ—à–µ–Ω–∏—è –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ –º—ã –ø–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ —Å—É–∂–∞–µ–º –¥–∏–∞–ø–∞–∑–æ–Ω –≤–æ–∑–º–æ–∂–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π $k$, –ø—Ä–æ–≤–µ—Ä—è—è –¥–ª—è –∫–∞–∂–¥–æ–≥–æ "—Å—Ä–µ–¥–Ω–µ–≥–æ" –∑–Ω–∞—á–µ–Ω–∏—è, –º–æ–∂–Ω–æ –ª–∏ –ø–æ–ª—É—á–∏—Ç—å $p$ –∏–∑ —Å—Ç—Ä–æ–∫–∏ $t$ –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —Å–∏–º–≤–æ–ª–æ–≤. –¢–∞–∫–æ–π –ø–æ–¥—Ö–æ–¥ –ø–æ–∑–≤–æ–ª—è–µ—Ç –±—ã—Å—Ç—Ä–æ –Ω–∞–π—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–¥–∞–ª–µ–Ω–∏–π, –ø–æ—Å–ª–µ –∫–æ—Ç–æ—Ä—ã—Ö —Å–ª–æ–≤–æ $p$ –æ—Å—Ç–∞–µ—Ç—Å—è –¥–æ—Å—Ç–∏–∂–∏–º—ã–º, —á—Ç–æ –∏ —è–≤–ª—è–µ—Ç—Å—è –∏—Å–∫–æ–º—ã–º —Ä–µ—à–µ–Ω–∏–µ–º –∑–∞–¥–∞—á–∏.

````python
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –≤—Ä–µ–º–µ–Ω–∏ - –û(TODO)
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –ø–∞–º—è—Ç–∏ - O(TODO)

# –í–ê–® –ö–û–î
import sys
from typing import List

def solve() -> None:
    """
    Description:
        –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏.
        –ß–∏—Ç–∞–µ—Ç –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ, –Ω–∞—Ö–æ–¥–∏—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ X,
        –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —É—Å–ª–æ–≤–∏–µ: sum(min(a[i], X)) >= X * k.

    Args:
        None

    Returns:
        None

    Raises:
        None

    Examples:
        >>> solve()
        # –í–≤–æ–¥: 3 4 1 2 3 4
        # –í—ã–≤–æ–¥: 2
    """
    data = sys.stdin.read().strip().split()
    if not data:
        return

    k = int(data[0])
    n = int(data[1])
    a = list(map(int, data[2:2 + n]))

    # –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–≤–µ—Ç–æ–≤ –Ω–µ –º–æ–∂–µ—Ç –ø—Ä–µ–≤—ã—à–∞—Ç—å sum(a) // k
    lo = 0
    hi = sum(a) // k
    ans = 0

    # –ë—É–¥–µ–º –∏—Å–∫–∞—Ç—å –ø–æ –¥–≤–æ–∏—á–Ω–æ–º—É –ø–æ–∏—Å–∫—É (–º–µ—Ç–æ–¥—É –¥–∏—Ö–æ—Ç–æ–º–∏–∏) –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ X,
    # –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —É—Å–ª–æ–≤–∏–µ: sum(min(a[i], X)) >= X * k.
    while lo <= hi:
        mid = (lo + hi) // 2
        total = 0
        for ai in a:
            total += min(ai, mid)
        if total >= mid * k:
            ans = mid
            lo = mid + 1
        else:
            hi = mid - 1

    sys.stdout.write(str(ans))

if __name__ == '__main__':
    solve()
````

–ó–∞–¥–∞—á–∞ —Å–≤–æ–¥–∏—Ç—Å—è –∫ –ø–æ–∏—Å–∫—É –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ —á–∏—Å–ª–∞ —Å–æ–≤–µ—Ç–æ–≤ $X$, –∫–æ—Ç–æ—Ä–æ–µ –º–æ–∂–Ω–æ —Å–æ—Å—Ç–∞–≤–∏—Ç—å, –µ—Å–ª–∏ –≤ –∫–∞–∂–¥–æ–º —Å–æ–≤–µ—Ç–µ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ä–æ–≤–Ω–æ $k$ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –∏–∑ —Ä–∞–∑–Ω—ã—Ö –≥—Ä—É–ø–ø. –î–ª—è –∫–∞–∂–¥–æ–≥–æ —Å–æ–≤–µ—Ç–∞ –∫–∞–∂–¥–∞—è –≥—Ä—É–ø–ø–∞ –º–æ–∂–µ—Ç –≤–Ω–µ—Å—Ç–∏ –Ω–µ –±–æ–ª–µ–µ –æ–¥–Ω–æ–≥–æ —Å—Ç—É–¥–µ–Ω—Ç–∞, –∞ –≥—Ä—É–ø–ø–∞ —Å $a[i]$ —Å—Ç—É–¥–µ–Ω—Ç–∞–º–∏ –º–æ–∂–µ—Ç —É—á–∞—Å—Ç–≤–æ–≤–∞—Ç—å –º–∞–∫—Å–∏–º—É–º $a[i]$ —Ä–∞–∑. –ü—Ä–∏ —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–∏ $X$ —Å–æ–≤–µ—Ç–æ–≤ –∫–∞–∂–¥–∞—è –≥—Ä—É–ø–ø–∞ —Å–ø–æ—Å–æ–±–Ω–∞ –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–∏—Ç—å –Ω–µ –±–æ–ª–µ–µ $\min(a[i], X)$ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —Å—É–º–º–∞—Ä–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤, –∫–æ—Ç–æ—Ä–æ–µ –º–æ–∂–Ω–æ –≤—ã–¥–µ–ª–∏—Ç—å –∏–∑ –≤—Å–µ—Ö –≥—Ä—É–ø–ø, —Ä–∞–≤–Ω–æ $\sum \min(a[i], X)$, –∏ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –Ω–µ –º–µ–Ω—å—à–µ $X \times k$.

–ò—Å–ø–æ–ª—å–∑—É—è –º–µ—Ç–æ–¥ –¥–∏—Ö–æ—Ç–æ–º–∏–∏ (–±–∏–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞), –º—ã –Ω–∞—Ö–æ–¥–∏–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ $X$, —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–µ–µ –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤—É $\sum_{i=1}^{n} \min(a[i], X) \ge X \times k$. –≠—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ –ø–æ–¥–æ–±—Ä–∞—Ç—å –Ω—É–∂–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ $X$ –¥–∞–∂–µ –ø—Ä–∏ –±–æ–ª—å—à–æ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –≤ –≥—Ä—É–ø–ø–∞—Ö.

### –í–ê–® —Å–∫—Ä–∏–Ω –û–ö–Ω—É—Ç–æ–π –∑–∞–¥–∞—á–∏

### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
- **–ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞** ‚Äî –Ω–µ–ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞ –¥–ª–∏–Ω–æ–π –¥–æ $100$, –æ–ø–∏—Å—ã–≤–∞—é—â–∞—è —Ä–µ—Ü–µ–ø—Ç –≥–∞–º–±—É—Ä–≥–µ—Ä–∞ ('B', 'S', 'C').  
- **–í—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∫–∞** ‚Äî —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ $n_b, n_s, n_c$ $(1 \leq n_b, n_s, n_c \leq 100)$ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ö–ª–µ–±–∞, –∫–æ–ª–±–∞—Å—ã –∏ —Å—ã—Ä–∞ –Ω–∞ –∫—É—Ö–Ω–µ.  
- **–¢—Ä–µ—Ç—å—è —Å—Ç—Ä–æ–∫–∞** ‚Äî —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ $p_b, p_s, p_c$ $(1 \leq p_b, p_s, p_c \leq 100)$ ‚Äî —Å—Ç–æ–∏–º–æ—Å—Ç—å –æ–¥–Ω–æ–≥–æ –∫—É—Å–æ—á–∫–∞ —Ö–ª–µ–±–∞, –∫–æ–ª–±–∞—Å—ã –∏ —Å—ã—Ä–∞ –≤ –º–∞–≥–∞–∑–∏–Ω–µ.  
- **–ß–µ—Ç–≤–µ—Ä—Ç–∞—è —Å—Ç—Ä–æ–∫–∞** ‚Äî –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ $r$ $(1 \leq r \leq 10^{12})$ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä—É–±–ª–µ–π —É –ü–æ–ª–∏–∫–∞—Ä–ø–∞.

### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
–í—ã–≤–µ–¥–∏—Ç–µ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äî –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–æ–≤, –∫–æ—Ç–æ—Ä–æ–µ –ü–æ–ª–∏–∫–∞—Ä–ø –º–æ–∂–µ—Ç –ø—Ä–∏–≥–æ—Ç–æ–≤–∏—Ç—å. –ï—Å–ª–∏ –Ω–µ —É–¥–∞—Å—Ç—Å—è –ø—Ä–∏–≥–æ—Ç–æ–≤–∏—Ç—å –Ω–∏ –æ–¥–Ω–æ–≥–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–∞, –≤—ã–≤–µ–¥–∏—Ç–µ `0`.

### –ü—Ä–∏–º–µ—Ä—ã  

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
BBBSSC
6 4 1
1 2 3
4
```
#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
2
```

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
BBC
1 10 1
1 10 1
21
```
#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
7
```

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
BSC
1 1 1
1 1 3
1000000000000
```
#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
200000000001
```

````python
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –≤—Ä–µ–º–µ–Ω–∏ - –û(TODO)
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –ø–∞–º—è—Ç–∏ - O(TODO)

# –í–ê–® –ö–û–î
import sys
from typing import Tuple

def solve() -> None:
    """
    Description:
        –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏.
        –ß–∏—Ç–∞–µ—Ç –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ, –≤—ã—á–∏—Å–ª—è–µ—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–æ–≤,
        –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–Ω–æ –ø—Ä–∏–≥–æ—Ç–æ–≤–∏—Ç—å, –Ω–µ –ø—Ä–µ–≤—ã—à–∞—è –±—é–¥–∂–µ—Ç r.

    Args:
        None

    Returns:
        None

    Raises:
        None

    Examples:
        >>> solve()
        # –í–≤–æ–¥: BSSC 1 2 3 10 20 30 100
        # –í—ã–≤–æ–¥: 3
    """
    input_data = sys.stdin.read().split()
    if not input_data:
        return

    # –ß—Ç–µ–Ω–∏–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
    recipe = input_data[0]
    n_b, n_s, n_c = map(int, input_data[1:4])
    p_b, p_s, p_c = map(int, input_data[4:7])
    r = int(input_data[7])

    # –ü–æ–¥—Å—á–∏—Ç–∞–µ–º —Ç—Ä–µ–±—É–µ–º–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –¥–ª—è –æ–¥–Ω–æ–≥–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–∞
    req_b = recipe.count('B')
    req_s = recipe.count('S')
    req_c = recipe.count('C')

    def cost(x: int) -> int:
        """
        Description:
            –í—ã—á–∏—Å–ª—è–µ—Ç —Å—É–º–º–∞—Ä–Ω—É—é —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–æ–∫—É–ø–∫–∏ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –¥–ª—è –∏–∑–≥–æ—Ç–æ–≤–ª–µ–Ω–∏—è x –≥–∞–º–±—É—Ä–≥–µ—Ä–æ–≤.

        Args:
            x: –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–æ–≤

        Returns:
            –°—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–æ–∫—É–ø–∫–∏ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤

        Raises:
            None

        Examples:
            >>> cost(3)
            60
        """
        # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç—ã, –∫–æ—Ç–æ—Ä—ã–µ –Ω—É–∂–Ω–æ –¥–æ–∫—É–ø–∏—Ç—å –¥–ª—è x –≥–∞–º–±—É—Ä–≥–µ—Ä–æ–≤
        buy_b = max(0, req_b * x - n_b)
        buy_s = max(0, req_s * x - n_s)
        buy_c = max(0, req_c * x - n_c)
        return buy_b * p_b + buy_s * p_s + buy_c * p_c

    # –ü—Ä–∏–º–µ–Ω—è–µ–º –º–µ—Ç–æ–¥ –¥–∏—Ö–æ—Ç–æ–º–∏–∏ (–±–∏–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞) –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ x,
    # —Ç–∞–∫–æ–≥–æ —á—Ç–æ —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–æ–∫—É–ø–∫–∏ –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç r.
    lo = 0
    hi = 10**13  # –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–ª—å—à–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –≤–µ—Ä—Ö–Ω–µ–π –≥—Ä–∞–Ω–∏—Ü—ã
    ans = 0
    while lo <= hi:
        mid = (lo + hi) // 2
        if cost(mid) <= r:
            ans = mid
            lo = mid + 1
        else:
            hi = mid - 1

    sys.stdout.write(str(ans))
````

### –í–ê–®–ê –∏–Ω—Ç—É–∏—Ü–∏—è

–†–µ—à–µ–Ω–∏–µ —Å–≤–æ–¥–∏—Ç—Å—è –∫ –ø–æ–∏—Å–∫—É –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≥–∞–º–±—É—Ä–≥–µ—Ä–æ–≤, –∫–æ—Ç–æ—Ä–æ–µ –º–æ–∂–Ω–æ –ø—Ä–∏–≥–æ—Ç–æ–≤–∏—Ç—å, –Ω–µ –ø–æ—Ç—Ä–∞—Ç–∏–≤ –±–æ–ª–µ–µ $ùëü$
r —Ä—É–±–ª–µ–π. –ó–¥–µ—Å—å –Ω–∞ –ø–æ–º–æ—â—å –ø—Ä–∏—Ö–æ–¥–∏—Ç –º–µ—Ç–æ–¥ –¥–∏—Ö–æ—Ç–æ–º–∏–∏ (–±–∏–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞), –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∑–≤–æ–ª—è–µ—Ç –∑–∞ –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–æ–µ —á–∏—Å–ª–æ —à–∞–≥–æ–≤ –Ω–∞–π—Ç–∏ —Ç–∞–∫–æ–µ —á–∏—Å–ª–æ, –ø—Ä–æ–≤–µ—Ä—è—è –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–∞–Ω–¥–∏–¥–∞—Ç–∞, —É–∫–ª–∞–¥—ã–≤–∞–µ—Ç—Å—è –ª–∏ —Å—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–æ–∫—É–ø–∫–∏ –≤ –∑–∞–¥–∞–Ω–Ω—ã–π –±—é–¥–∂–µ—Ç. –ú—ã –ø–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ —Å—É–∂–∞–µ–º –¥–∏–∞–ø–∞–∑–æ–Ω –≤–æ–∑–º–æ–∂–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π, –æ–ø—Ä–µ–¥–µ–ª—è—è, —Å–∫–æ–ª—å–∫–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–æ–≤ —Ä–µ–∞–ª—å–Ω–æ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å, –ø—Ä–µ–∂–¥–µ —á–µ–º –ø—Ä–µ–≤—ã—à–∞–µ—Ç—Å—è –ª–∏–º–∏—Ç –¥–µ–Ω–µ–≥.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –∏—Å–ø–æ–ª—å–∑—É—è –±–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫, –º—ã —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ –ø–æ–¥–±–∏—Ä–∞–µ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–æ–≤, –∫–æ—Ç–æ—Ä–æ–µ –ü–æ–ª–∏–∫–∞—Ä–ø —Å–º–æ–∂–µ—Ç –ø—Ä–∏–≥–æ—Ç–æ–≤–∏—Ç—å, –Ω–µ –≤—ã—Ö–æ–¥—è –∑–∞ —Ä–∞–º–∫–∏ –¥–æ—Å—Ç—É–ø–Ω–æ–≥–æ –±—é–¥–∂–µ—Ç–∞. –¢–∞–∫–æ–π –ø–æ–¥—Ö–æ–¥ –ø–æ–∑–≤–æ–ª—è–µ—Ç —É—á–µ—Å—Ç—å –∫–∞–∫ –∏–º–µ—é—â–∏–µ—Å—è –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç—ã, —Ç–∞–∫ –∏ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –¥–æ–∫—É–ø–∫–∏, –æ–±–µ—Å–ø–µ—á–∏–≤–∞—è —Ç–æ—á–Ω–æ–µ –∏ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏.

### –í–ê–® —Å–∫—Ä–∏–Ω –û–ö–Ω—É—Ç–æ–π –∑–∞–¥–∞—á–∏

## –ó–∞–¥–∞—á–∞ –°–æ–±—Ä–∞—Ç—å—Å—è –≤–º–µ—Å—Ç–µ ([—Å—Å—ã–ª–∫–∞](https://codeforces.com/edu/course/2/lesson/6/3/practice/contest/285083/problem/A)) [0.25 –±–∞–ª–ª–∞]

–ù–∞ –ø—Ä—è–º–æ–π —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω—ã $n$ —á–µ–ª–æ–≤–µ–∫. –ö–∞–∂–¥—ã–π —á–µ–ª–æ–≤–µ–∫ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –Ω–µ–∫–æ—Ç–æ—Ä–æ–π —Ç–æ—á–∫–µ $x_i$ –∏ –º–æ–∂–µ—Ç –¥–≤–∏–≥–∞—Ç—å—Å—è —Å–æ —Å–∫–æ—Ä–æ—Å—Ç—å—é $v_i$. –ò–º –Ω—É–∂–Ω–æ —Å–æ–±—Ä–∞—Ç—å—Å—è –≤ –æ–¥–Ω–æ–π —Ç–æ—á–∫–µ –∫–∞–∫ –º–æ–∂–Ω–æ –±—ã—Å—Ç—Ä–µ–µ. –¢—Ä–µ–±—É–µ—Ç—Å—è –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è, –∑–∞ –∫–æ—Ç–æ—Ä–æ–µ –æ–Ω–∏ –º–æ–≥—É—Ç —ç—Ç–æ–≥–æ –¥–æ–±–∏—Ç—å—Å—è.  

### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
- **–ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞** —Å–æ–¥–µ—Ä–∂–∏—Ç –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ $n$ $(1 \leq n \leq 10^5)$ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª—é–¥–µ–π.  
- **–°–ª–µ–¥—É—é—â–∏–µ $n$ —Å—Ç—Ä–æ–∫** —Å–æ–¥–µ—Ä–∂–∞—Ç –ø–æ –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ $x_i$ –∏ $v_i$ $(-10^9 \leq x_i \leq 10^9, 1 \leq v_i \leq 10^9)$ ‚Äî –Ω–∞—á–∞–ª—å–Ω–æ–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ —á–µ–ª–æ–≤–µ–∫–∞ –∏ –µ–≥–æ —Å–∫–æ—Ä–æ—Å—Ç—å.  

### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
–í—ã–≤–µ–¥–∏—Ç–µ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äî –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è, –∑–∞ –∫–æ—Ç–æ—Ä–æ–µ –≤—Å–µ –ª—é–¥–∏ –º–æ–≥—É—Ç —Å–æ–±—Ä–∞—Ç—å—Å—è –≤ –æ–¥–Ω–æ–π —Ç–æ—á–∫–µ. –û—Ç–≤–µ—Ç –±—É–¥–µ—Ç —Å—á–∏—Ç–∞—Ç—å—Å—è –≤–µ—Ä–Ω—ã–º, –µ—Å–ª–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –∏–ª–∏ –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç $10^{-6}$.  

### –ü—Ä–∏–º–µ—Ä—ã  

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
5
-1 5
10 3
4 2
7 10
8 1
```
#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
1.5
```

````python
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –≤—Ä–µ–º–µ–Ω–∏ - –û(TODO)
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –ø–∞–º—è—Ç–∏ - O(TODO)

# –í–ê–® –ö–û–î
import sys
from typing import List, Tuple

def solve() -> None:
    """
    Description:
        –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏.
        –ß–∏—Ç–∞–µ—Ç –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ, –≤—ã—á–∏—Å–ª—è–µ—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è t,
        –∑–∞ –∫–æ—Ç–æ—Ä–æ–µ –≤—Å–µ –ª—é–¥–∏ –º–æ–≥—É—Ç —Å–æ–±—Ä–∞—Ç—å—Å—è –≤ –æ–¥–Ω–æ–π —Ç–æ—á–∫–µ.

    Args:
        None

    Returns:
        None

    Raises:
        None

    Examples:
        >>> solve()
        # –í–≤–æ–¥: 3 1.0 2.0 2.0 3.0 3.0 4.0
        # –í—ã–≤–æ–¥: 0.500000
    """
    input_data = sys.stdin.read().split()
    if not input_data:
        return

    n = int(input_data[0])
    people: List[Tuple[float, float]] = []
    idx = 1
    for _ in range(n):
        x = float(input_data[idx])
        v = float(input_data[idx + 1])
        idx += 2
        people.append((x, v))

    def can_meet(t: float) -> bool:
        """
        Description:
            –ü—Ä–æ–≤–µ—Ä—è–µ—Ç, –º–æ–∂–Ω–æ –ª–∏ –∑–∞ –≤—Ä–µ–º—è t –≤—ã–±—Ä–∞—Ç—å —Ç–æ—á–∫—É,
            –¥–æ –∫–æ—Ç–æ—Ä–æ–π —É—Å–ø–µ—é—Ç –¥–æ–±—Ä–∞—Ç—å—Å—è –≤—Å–µ –ª—é–¥–∏.

        Args:
            t: –í—Ä–µ–º—è, –∑–∞ –∫–æ—Ç–æ—Ä–æ–µ –≤—Å–µ –ª—é–¥–∏ –¥–æ–ª–∂–Ω—ã —Å–æ–±—Ä–∞—Ç—å—Å—è –≤ –æ–¥–Ω–æ–π —Ç–æ—á–∫–µ

        Returns:
            True, –µ—Å–ª–∏ –º–æ–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å —Ç–∞–∫—É—é —Ç–æ—á–∫—É, –∏–Ω–∞—á–µ False

        Raises:
            None

        Examples:
            >>> can_meet(0.5)
            True
        """
        left_bound = -1e18
        right_bound = 1e18
        for x, v in people:
            left_bound = max(left_bound, x - v * t)
            right_bound = min(right_bound, x + v * t)
        return left_bound <= right_bound

    lo = 0.0
    hi = 1e10      # –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–ª—å—à–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –≤–µ—Ä—Ö–Ω–µ–π –≥—Ä–∞–Ω–∏—Ü—ã –≤—Ä–µ–º–µ–Ω–∏
    for _ in range(100):  # 100 –∏—Ç–µ—Ä–∞—Ü–∏–π –¥–ª—è –¥–æ—Å—Ç–∏–∂–µ–Ω–∏—è –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–π —Ç–æ—á–Ω–æ—Å—Ç–∏
        mid = (lo + hi) / 2.0
        if can_meet(mid):
            hi = mid
        else:
            lo = mid

    # –í—ã–≤–æ–¥ –æ—Ç–≤–µ—Ç–∞ –≤ –≤–∏–¥–µ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞ —Å 6 –∑–Ω–∞–∫–∞–º–∏ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
    print("{:.6f}".format(hi))
````

### –í–ê–®–ê –∏–Ω—Ç—É–∏—Ü–∏—è

–î–ª—è –∫–∞–∂–¥–æ–≥–æ —á–µ–ª–æ–≤–µ–∫–∞ –∑–∞–¥–∞—ë—Ç—Å—è –∏–Ω—Ç–µ—Ä–≤–∞–ª, –≤ –∫–æ—Ç–æ—Ä–æ–º –æ–Ω –º–æ–∂–µ—Ç –æ–∫–∞–∑–∞—Ç—å—Å—è –∑–∞ –≤—Ä–µ–º—è $(t)$: $([x_i - v_i \cdot t,\, x_i + v_i \cdot t])$. –ï—Å–ª–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ç–∞–∫–∞—è —Ç–æ—á–∫–∞, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—é –≤—Å–µ—Ö —ç—Ç–∏—Ö –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤, —Ç–æ –∑–∞ –≤—Ä–µ–º—è $(t)$ –ª—é–¥–∏ –º–æ–≥—É—Ç —Å–æ–±—Ä–∞—Ç—å—Å—è –≤–º–µ—Å—Ç–µ. –ë–ª–∞–≥–æ–¥–∞—Ä—è –º–æ–Ω–æ—Ç–æ–Ω–Ω–æ—Å—Ç–∏ –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ –ø–æ $(t)$ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥ –¥–∏—Ö–æ—Ç–æ–º–∏–∏ (–±–∏–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞) –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ $(t)$, —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–µ–≥–æ —ç—Ç–æ–º—É —É—Å–ª–æ–≤–∏—é.

### –í–ê–® —Å–∫—Ä–∏–Ω –û–ö–Ω—É—Ç–æ–π –∑–∞–¥–∞—á–∏

## –ó–∞–¥–∞—á–∞ –ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —Å—Ä–µ–¥–Ω–∏–π –æ—Ç—Ä–µ–∑–æ–∫ ([—Å—Å—ã–ª–∫–∞](https://codeforces.com/edu/course/2/lesson/6/4/practice/contest/285069/problem/A)) [0.25 –±–∞–ª–ª–∞]

–î–∞–Ω –º–∞—Å—Å–∏–≤ –∏–∑ $n$ —á–∏—Å–µ–ª –∏ —á–∏—Å–ª–æ $d$. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏ —Ç–∞–∫–æ–π –æ—Ç—Ä–µ–∑–æ–∫ –º–∞—Å—Å–∏–≤–∞ –¥–ª–∏–Ω—ã –Ω–µ –º–µ–Ω–µ–µ $d$, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ.  

### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
- **–ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞** —Å–æ–¥–µ—Ä–∂–∏—Ç –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ $n$ –∏ $d$ $(1 \leq d \leq n \leq 10^5)$ ‚Äî –¥–ª–∏–Ω–∞ –º–∞—Å—Å–∏–≤–∞ –∏ –º–∏–Ω–∏–º–∞–ª—å–Ω–∞—è –¥–ª–∏–Ω–∞ –æ—Ç—Ä–µ–∑–∫–∞.  
- **–í—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∫–∞** —Å–æ–¥–µ—Ä–∂–∏—Ç $n$ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª $a_i$ $(0 \leq a_i \leq 100)$ ‚Äî —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Å—Å–∏–≤–∞.  

### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
–í—ã–≤–µ–¥–∏—Ç–µ –¥–≤–∞ —á–∏—Å–ª–∞ ‚Äî –Ω–æ–º–µ—Ä–∞ –ø–µ—Ä–≤–æ–≥–æ –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ –Ω–∞–π–¥–µ–Ω–Ω–æ–≥–æ –æ—Ç—Ä–µ–∑–∫–∞ (–Ω—É–º–µ—Ä–∞—Ü–∏—è —Å $1$).  

### –ü—Ä–∏–º–µ—Ä—ã  

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
6 2
3 1 8 5 7 2
```  
#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
3 5
```

````python
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –≤—Ä–µ–º–µ–Ω–∏ - –û(TODO)
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –ø–∞–º—è—Ç–∏ - O(TODO)

# –í–ê–® –ö–û–î
# –ò–º–ø–æ—Ä—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import sys
from typing import List, Tuple


def solve() -> None:
    """
    Description:
        –†–µ—à–∞–µ—Ç –∑–∞–¥–∞—á—É –ø–æ–∏—Å–∫–∞ –Ω–∞–∏–±–æ–ª—å—à–µ–≥–æ –ø–æ–¥–æ—Ç—Ä–µ–∑–∫–∞ –º–∞—Å—Å–∏–≤–∞, —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–æ—Ç–æ—Ä–æ–≥–æ
        –Ω–µ –º–µ–Ω—å—à–µ –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –ø–æ—Ä–æ–≥–∞, —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –º–µ—Ç–æ–¥–∞ –±–∏–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞.

    Args:
        –î–∞–Ω–Ω—ã–µ —Å—á–∏—Ç—ã–≤–∞—é—Ç—Å—è –∏–∑ `sys.stdin`, –≤–∫–ª—é—á–∞—è:
        - –ü–µ—Ä–≤—É—é —Å—Ç—Ä–æ–∫—É: –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ `n` (—Ä–∞–∑–º–µ—Ä –º–∞—Å—Å–∏–≤–∞) –∏ `d` (–º–∏–Ω–∏–º–∞–ª—å–Ω–∞—è –¥–ª–∏–Ω–∞ –ø–æ–¥–æ—Ç—Ä–µ–∑–∫–∞).
        - –í—Ç–æ—Ä—É—é —Å—Ç—Ä–æ–∫—É: –º–∞—Å—Å–∏–≤ `a` –∏–∑ `n` —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª.

    Returns:
        –í—ã–≤–æ–¥–∏—Ç –≤ `sys.stdout` –¥–≤–∞ —á–∏—Å–ª–∞ ‚Äî –∏–Ω–¥–µ–∫—Å—ã (1-–∏–Ω–¥–µ–∫—Å–∞—Ü–∏—è) –ª–µ–≤–æ–≥–æ –∏ –ø—Ä–∞–≤–æ–≥–æ –∫–æ–Ω—Ü–æ–≤
        –Ω–∞–π–¥–µ–Ω–Ω–æ–≥–æ –ø–æ–¥–æ—Ç—Ä–µ–∑–∫–∞.

    Raises:
        ValueError: –ï—Å–ª–∏ –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã.

    Examples:
        –í—Ö–æ–¥:
            5 2
            1 2 3 4 5
        –í—ã—Ö–æ–¥:
            4 5
    """

    # –ß—Ç–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–∏ —Å n –∏ d
    line = sys.stdin.readline()
    if not line:
        return

    n, d = map(int, line.strip().split())

    # –ß—Ç–µ–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ —á–∏—Å–µ–ª
    a = list(map(int, sys.stdin.readline().strip().split()))

    # –ì—Ä–∞–Ω–∏—Ü—ã –±–∏–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞ (–¥–∏–∞–ø–∞–∑–æ–Ω –≤–æ–∑–º–æ–∂–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π)
    lo, hi = 0.0, 100.0

    # –ò–∑–Ω–∞—á–∞–ª—å–Ω–æ –ª—É—á—à–∏–π –Ω–∞–π–¥–µ–Ω–Ω—ã–π –æ—Ç—Ä–µ–∑–æ–∫ ‚Äî [1, d]
    best_l, best_r = 1, d

    # –ü—Ä–µ—Ñ–∏–∫—Å–Ω—ã–µ —Å—É–º–º—ã –¥–ª—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞
    prefix: List[float] = [0.0] * (n + 1)

    # –ë–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫ —Å 60 –∏—Ç–µ—Ä–∞—Ü–∏—è–º–∏ (—Ç–æ—á–Ω–æ—Å—Ç—å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–∞)
    for _ in range(60):
        mid = (lo + hi) / 2.0  # –°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ

        prefix[0] = 0.0

        # –í—ã—á–∏—Å–ª—è–µ–º –ø—Ä–µ—Ñ–∏–∫—Å–Ω—ã–µ —Å—É–º–º—ã –¥–ª—è –º–∞—Å—Å–∏–≤–∞, –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω–æ–≥–æ –ø–æ `a[i] - mid`
        for i in range(1, n + 1):
            prefix[i] = prefix[i - 1] + (a[i - 1] - mid)

        # –§–ª–∞–≥, –æ–±–æ–∑–Ω–∞—á–∞—é—â–∏–π, –Ω–∞—à–ª–∏ –ª–∏ –º—ã –ø–æ–¥—Ö–æ–¥—è—â–∏–π –ø–æ–¥–æ—Ç—Ä–µ–∑–æ–∫
        feasible = False

        # –ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è –ø—Ä–µ—Ñ–∏–∫—Å–Ω–∞—è —Å—É–º–º–∞ –∏ –µ–µ –∏–Ω–¥–µ–∫—Å
        min_prefix = 0.0
        min_index = 0

        # –ü–µ—Ä–µ–±–∏—Ä–∞–µ–º –≤–æ–∑–º–æ–∂–Ω—ã–µ –ø—Ä–∞–≤—ã–µ –≥—Ä–∞–Ω–∏—Ü—ã –ø–æ–¥–æ—Ç—Ä–µ–∑–∫–∞
        for i in range(d, n + 1):
            # –û–±–Ω–æ–≤–ª—è–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –ø—Ä–µ—Ñ–∏–∫—Å–Ω—É—é —Å—É–º–º—É
            if prefix[i - d] < min_prefix:
                min_prefix = prefix[i - d]
                min_index = i - d

            # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ª–∏ –ø–æ–¥–æ—Ç—Ä–µ–∑–æ–∫ –¥–ª–∏–Ω—ã >= d —Å –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π —Å—É–º–º–æ–π
            if prefix[i] - min_prefix >= 0:
                feasible = True
                current_l = min_index + 1  # –ü–µ—Ä–µ—Ö–æ–¥ –∫ 1-–∏–Ω–¥–µ–∫—Å–∞—Ü–∏–∏
                current_r = i
                break

        # –û–±–Ω–æ–≤–ª—è–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –±–∏–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞
        if feasible:
            lo = mid
            best_l, best_r = current_l, current_r
        else:
            hi = mid

    # –í—ã–≤–æ–¥–∏–º –Ω–∞–π–¥–µ–Ω–Ω—ã–π –ø–æ–¥–æ—Ç—Ä–µ–∑–æ–∫
    sys.stdout.write(f"{best_l} {best_r}\n")
````

### –í–ê–®–ê –∏–Ω—Ç—É–∏—Ü–∏—è

–ù–∞–º –¥–∞–Ω –º–∞—Å—Å–∏–≤ —á–∏—Å–µ–ª, –∏ —Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–∞–π—Ç–∏ –æ—Ç—Ä–µ–∑–æ–∫ –¥–ª–∏–Ω–æ–π –Ω–µ –º–µ–Ω—å—à–µ $(d)$, —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∫–æ—Ç–æ—Ä–æ–≥–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ. –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö –æ—Ç—Ä–µ–∑–∫–æ–≤ –Ω—É–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å —Ç–æ—Ç, —É –∫–æ—Ç–æ—Ä–æ–≥–æ —Å—É–º–º–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –¥–µ–ª—ë–Ω–Ω–∞—è –Ω–∞ –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ, –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞.

–ß—Ç–æ–±—ã —Ä–µ—à–∏—Ç—å –∑–∞–¥–∞—á—É, –º—ã –ø—Ä–∏–º–µ–Ω—è–µ–º –º–µ—Ç–æ–¥ –±–∏–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞ –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é —Å—Ä–µ–¥–Ω–µ–≥–æ $(m)$. –ò–¥–µ—è –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ–±—ã –∑–∞–¥–∞–≤–∞—Ç—å –≤–æ–ø—Ä–æ—Å: ¬´–°—É—â–µ—Å—Ç–≤—É–µ—Ç –ª–∏ –æ—Ç—Ä–µ–∑–æ–∫, —Å—Ä–µ–¥–Ω–µ–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –Ω–µ –º–µ–Ω—å—à–µ $(m)$?¬ª –ú—ã –∑–Ω–∞–µ–º, —á—Ç–æ –≤–æ–∑–º–æ–∂–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –ª–µ–∂–∏—Ç –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –æ—Ç 0 –¥–æ 100, –ø–æ—ç—Ç–æ–º—É –º–æ–∂–µ–º –Ω–∞—á–∞—Ç—å —Å —ç—Ç–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –∏ –ø–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ –µ–≥–æ —Å—É–∂–∞—Ç—å.

–î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∫–∞–Ω–¥–∏–¥–∞—Ç–∞ $(m)$ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç –º–∞—Å—Å–∏–≤–∞, –≤—ã—á–∏—Ç–∞—è $(m)$ –∏–∑ –Ω–µ–≥–æ. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –µ—Å–ª–∏ –∏—Å—Ö–æ–¥–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –±–æ–ª—å—à–µ –∏–ª–∏ —Ä–∞–≤–µ–Ω $(m)$, —Ä–∞–∑–Ω–æ—Å—Ç—å –±—É–¥–µ—Ç –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π, –∞ –µ—Å–ª–∏ –º–µ–Ω—å—à–µ ‚Äì –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π. –ï—Å–ª–∏ –¥–ª—è –∫–∞–∫–æ–≥–æ-–ª–∏–±–æ –æ—Ç—Ä–µ–∑–∫–∞ —Å—É–º–º–∞ —ç—Ç–∏—Ö –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –æ–∫–∞–∂–µ—Ç—Å—è –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π, —Ç–æ –∏—Å—Ö–æ–¥–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ —ç—Ç–æ–≥–æ –æ—Ç—Ä–µ–∑–∫–∞ –±—É–¥–µ—Ç –Ω–µ –º–µ–Ω—å—à–µ $(m)$.

–ß—Ç–æ–±—ã —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ –≤—ã—á–∏—Å–ª—è—Ç—å —Å—É–º–º—É –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–≥–æ –æ—Ç—Ä–µ–∑–∫–∞, –∏—Å–ø–æ–ª—å–∑—É–µ–º –ø—Ä–µ—Ñ–∏–∫—Å–Ω—ã–µ —Å—É–º–º—ã, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–∑–≤–æ–ª—è—é—Ç –±—ã—Å—Ç—Ä–æ –Ω–∞—Ö–æ–¥–∏—Ç—å —Å—É–º–º—É —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –æ—Ç—Ä–µ–∑–∫–∞ –∫–∞–∫ —Ä–∞–∑–Ω–∏—Ü—É –Ω–∞–∫–æ–ø–ª–µ–Ω–Ω—ã—Ö –∏—Ç–æ–≥–æ–≤. –° –ø–æ–º–æ—â—å—é —ç—Ç–æ–≥–æ –º–µ—Ç–æ–¥–∞ –º—ã –ø–µ—Ä–µ–±–∏—Ä–∞–µ–º –≤–æ–∑–º–æ–∂–Ω—ã–µ –æ—Ç—Ä–µ–∑–∫–∏ –¥–ª–∏–Ω—ã –Ω–µ –º–µ–Ω—å—à–µ $(d)$ –∏, –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –Ω–∞–ª–∏—á–∏—è –ø–æ–¥—Ö–æ–¥—è—â–µ–≥–æ –æ—Ç—Ä–µ–∑–∫–∞, –∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä—É–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –±–∏–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞, –ø–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ –ø—Ä–∏–±–ª–∏–∂–∞—è—Å—å –∫ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–º—É –≤–æ–∑–º–æ–∂–Ω–æ–º—É —Å—Ä–µ–¥–Ω–µ–º—É –∏ –Ω–∞—Ö–æ–¥—è –Ω—É–∂–Ω—ã–π –æ—Ç—Ä–µ–∑–æ–∫.

## –ó–∞–¥–∞—á–∞: –í—ã–±–æ—Ä –ø–∞—Ä ([—Å—Å—ã–ª–∫–∞](https://codeforces.com/edu/course/2/lesson/6/4/practice/contest/285069/problem/C)) [1/3 –±–∞–ª–ª–∞]

–î–∞–Ω–æ $n$ –ø–∞—Ä –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª $(a_1, b_1), (a_2, b_2), \dots, (a_n, b_n)$. –¢—Ä–µ–±—É–µ—Ç—Å—è –≤—ã–±—Ä–∞—Ç—å $k$ –∏–∑ –Ω–∏—Ö $(1 \leq k \leq n)$ —Ç–∞–∫, —á—Ç–æ–±—ã –æ—Ç–Ω–æ—à–µ–Ω–∏–µ  

$\frac{a_{i_1} + a_{i_2} + \dots + a_{i_k}}{b_{i_1} + b_{i_2} + \dots + b_{i_k}}$

–±—ã–ª–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –≤–æ–∑–º–æ–∂–Ω—ã–º.  

### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
- **–ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞** —Å–æ–¥–µ—Ä–∂–∏—Ç –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ $n$ –∏ $k$ $(1 \leq k \leq n \leq 10^5)$ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä, –∫–æ—Ç–æ—Ä—ã–µ –Ω—É–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å.  
- **–°–ª–µ–¥—É—é—â–∏–µ $n$ —Å—Ç—Ä–æ–∫** —Å–æ–¥–µ—Ä–∂–∞—Ç –ø–æ –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ $a_i, b_i$ $(1 \leq a_i, b_i \leq 10^5)$ ‚Äî –∑–Ω–∞—á–µ–Ω–∏—è –∫–∞–∂–¥–æ–π –ø–∞—Ä—ã.  

### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
–í—ã–≤–µ–¥–∏—Ç–µ –∏—Å–∫–æ–º–æ–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ. –û—Ç–≤–µ—Ç –±—É–¥–µ—Ç —Å—á–∏—Ç–∞—Ç—å—Å—è –≤–µ—Ä–Ω—ã–º, –µ—Å–ª–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –∏–ª–∏ –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç $10^{-6}$.

````python
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –≤—Ä–µ–º–µ–Ω–∏ - –û(TODO)
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –ø–∞–º—è—Ç–∏ - O(TODO)

# –í–ê–® –ö–û–î
# –ò–º–ø–æ—Ä—Ç—ã —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import sys
import heapq

def main() -> None:
    """
    Description:
        –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–æ–≥—Ä–∞–º–º—ã. –°—á–∏—Ç—ã–≤–∞–µ—Ç –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ, –≤—ã–ø–æ–ª–Ω—è–µ—Ç –±–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫
        –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –æ—Ç–Ω–æ—à–µ–Ω–∏—è a_i / b_i –∏ –≤—ã–≤–æ–¥–∏—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

    Args:
        –ù–µ—Ç –≤—Ö–æ–¥–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.

    Returns:
        None

    Raises:
        –ù–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏–π.

    Examples:
        –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è:
        $ echo "3 2 1 2 3 4 5 6" | python script.py
        1.5000000000
    """
    data = sys.stdin.read().strip().split()

    # –ï—Å–ª–∏ –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç, –∑–∞–≤–µ—Ä—à–∞–µ–º —Ä–∞–±–æ—Ç—É
    if not data:
        return

    n = int(data[0])
    k = int(data[1])
    pairs = []

    # –§–æ—Ä–º–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫ –ø–∞—Ä (a, b)
    for i in range(n):
        a = float(data[2 + 2 * i])
        b = float(data[2 + 2 * i + 1])
        pairs.append((a, b))

    lo = 0.0
    hi = 100000.0  # –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ: a_i / b_i –º–æ–∂–µ—Ç –±—ã—Ç—å –Ω–µ –±–æ–ª—å—à–µ 100000 / 1 = 100000
    eps = 1e-7     # –¢–æ—á–Ω–æ—Å—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω–∏–π

    # –ë–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫ –ø–æ –∏—Å–∫–æ–º–æ–º—É –æ—Ç–Ω–æ—à–µ–Ω–∏—é
    while hi - lo > eps:
        mid = (lo + hi) / 2
        # –î–ª—è –∫–∞–∂–¥–æ–π –ø–∞—Ä—ã –≤—ã—á–∏—Å–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ a - mid * b
        v = [a - mid * b for a, b in pairs]
        # –í—ã–±–∏—Ä–∞–µ–º k –ø–∞—Ä —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –∏ –ø—Ä–æ–≤–µ—Ä—è–µ–º –∏—Ö —Å—É–º–º—É
        largest = heapq.nlargest(k, v)
        if sum(largest) >= 0:
            lo = mid
        else:
            hi = mid

    sys.stdout.write("{:.10f}".format(lo))
````

### –í–ê–®–ê –∏–Ω—Ç—É–∏—Ü–∏—è

–∞–¥–∞—á–∞ —Å–≤–æ–¥–∏—Ç—Å—è –∫ –≤—ã–±–æ—Ä—É $k$ –ø–∞—Ä –∏–∑ $n$ —Ç–∞–∫, —á—Ç–æ–±—ã –º–∞–∫—Å–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å –æ—Ç–Ω–æ—à–µ–Ω–∏–µ —Å—É–º–º $ \frac{a_1 + a_2 + \dots}{b_1 + b_2 + \dots} $. –ù–∞–º –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —ç—Ç–æ–≥–æ –æ—Ç–Ω–æ—à–µ–Ω–∏—è.

–ö–ª—é—á–µ–≤–∞—è –∏–¥–µ—è —Ä–µ—à–µ–Ω–∏—è ‚Äî –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å, —á—Ç–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ $R$, –∏ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å, –º–æ–∂–Ω–æ –ª–∏ –≤—ã–±—Ä–∞—Ç—å $k$ –ø–∞—Ä —Ç–∞–∫, —á—Ç–æ–±—ã —Å—É–º–º–∞ $a_i - R \cdot b_i$ –¥–ª—è –≤—ã–±—Ä–∞–Ω–Ω—ã—Ö –ø–∞—Ä –±—ã–ª–∞ –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π. –ï—Å–ª–∏ –º–æ–∂–Ω–æ, –∑–Ω–∞—á–∏—Ç, –æ—Ç–Ω–æ—à–µ–Ω–∏–µ $R$ –¥–æ—Å—Ç–∏–∂–∏–º–æ.

–î–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è —Ç–∞–∫–æ–≥–æ $R$ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –±–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫: –º—ã –∑–∞–¥–∞—ë–º –¥–∏–∞–ø–∞–∑–æ–Ω –≤–æ–∑–º–æ–∂–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π $R$ –∏ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–≤–µ—Ä—è–µ–º —Å–µ—Ä–µ–¥–∏–Ω—É —ç—Ç–æ–≥–æ –¥–∏–∞–ø–∞–∑–æ–Ω–∞. –ï—Å–ª–∏ –¥–ª—è –≤—ã–±—Ä–∞–Ω–Ω—ã—Ö $k$ –ø–∞—Ä —Å—É–º–º–∞ $a_i - R \cdot b_i$ –æ–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è –Ω–µ–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π, –º—ã –º–æ–∂–µ–º –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å —É–≤–µ–ª–∏—á–∏—Ç—å $R$; –µ—Å–ª–∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π ‚Äî —É–º–µ–Ω—å—à–∞–µ–º $R$.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –º–µ—Ç–æ–¥ –±–∏–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞ –ø–æ–∑–≤–æ–ª—è–µ—Ç —à–∞–≥ –∑–∞ —à–∞–≥–æ–º –ø—Ä–∏–±–ª–∏–∂–∞—Ç—å—Å—è –∫ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–º—É –æ—Ç–Ω–æ—à–µ–Ω–∏—é —Å –Ω—É–∂–Ω–æ–π —Ç–æ—á–Ω–æ—Å—Ç—å—é. –î–ª—è –∫–∞–∂–¥–æ–π –ø—Ä–æ–≤–µ—Ä–∫–∏ –≤—ã–±–∏—Ä–∞—é—Ç—Å—è $k$ –ø–∞—Ä —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ $a_i - R \cdot b_i$, —á—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –º–æ–∂–Ω–æ –ª–∏ –¥–æ–±–∏—Ç—å—Å—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –æ—Ç–Ω–æ—à–µ–Ω–∏—è.

### –í–ê–® —Å–∫—Ä–∏–Ω –û–ö–Ω—É—Ç–æ–π –∑–∞–¥–∞—á–∏

## –ó–∞–¥–∞—á–∞: –¢–∞–±–ª–∏—Ü–∞ —É–º–Ω–æ–∂–µ–Ω–∏—è ([—Å—Å—ã–ª–∫–∞](https://codeforces.com/edu/course/2/lesson/6/5/practice/contest/285084/problem/B)) [1/3 –±–∞–ª–ª–∞]  

–ü–µ—Ç—è —Å–æ—Å—Ç–∞–≤–∏–ª —Ç–∞–±–ª–∏—Ü—É —É–º–Ω–æ–∂–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–æ–º $n \times n$. –Ø—á–µ–π–∫–∞ –≤ $i$-–π —Å—Ç—Ä–æ–∫–µ –∏ $j$-–º —Å—Ç–æ–ª–±—Ü–µ —Å–æ–¥–µ—Ä–∂–∏—Ç –∑–Ω–∞—á–µ–Ω–∏–µ $i \cdot j$. –ü–µ—Ç—é –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç, –∫–∞–∫–æ–µ —á–∏—Å–ª–æ –≤ —Ç–∞–±–ª–∏—Ü–µ —è–≤–ª—è–µ—Ç—Å—è $k$-–º –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é. –ü–æ–º–æ–≥–∏—Ç–µ –µ–º—É –Ω–∞–π—Ç–∏ –æ—Ç–≤–µ—Ç.  

### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
- **–û–¥–Ω–∞ —Å—Ç—Ä–æ–∫–∞** —Å–æ–¥–µ—Ä–∂–∏—Ç –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ $n$ –∏ $k$ $(1 \leq n \leq 10^5, 1 \leq k \leq n^2)$.  

### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
–í—ã–≤–µ–¥–∏—Ç–µ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äî $k$-–µ —á–∏—Å–ª–æ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é –≤ —Ç–∞–±–ª–∏—Ü–µ.  

### –ü—Ä–∏–º–µ—Ä—ã  

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

```  
#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

```  
#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

````python
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –≤—Ä–µ–º–µ–Ω–∏ - –û(TODO)
# –ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∞—è —Å–ª–æ–∂–Ω–æ—Å—Ç—å –ø–æ –ø–∞–º—è—Ç–∏ - O(TODO)

# –í–ê–® –ö–û–î
from typing import List

def count_less_equal(n: int, mid: int) -> int:
    """
    Description:
        –ü–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª, –∫–æ—Ç–æ—Ä—ã–µ –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω—ã mid –≤ –º–∞—Ç—Ä–∏—Ü–µ n x n.

    Args:
        n: –†–∞–∑–º–µ—Ä –º–∞—Ç—Ä–∏—Ü—ã (n x n)
        mid: –ó–Ω–∞—á–µ–Ω–∏–µ, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ –ø–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª

    Returns:
        –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª, –∫–æ—Ç–æ—Ä—ã–µ –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω—ã mid

    Examples:
        >>> count_less_equal(3, 5)
        6
    """
    count = 0
    for i in range(1, min(n, mid) + 1):
        count += min(mid // i, n)
    return count

def find_kth_number(n: int, k: int) -> int:
    """
    Description:
        –ù–∞—Ö–æ–¥–∏—Ç k-–µ –ø–æ –≤–µ–ª–∏—á–∏–Ω–µ —á–∏—Å–ª–æ –≤ –º–∞—Ç—Ä–∏—Ü–µ n x n.

    Args:
        n: –†–∞–∑–º–µ—Ä –º–∞—Ç—Ä–∏—Ü—ã (n x n)
        k: –ü–æ—Ä—è–¥–∫–æ–≤—ã–π –Ω–æ–º–µ—Ä —á–∏—Å–ª–∞

    Returns:
        k-–µ –ø–æ –≤–µ–ª–∏—á–∏–Ω–µ —á–∏—Å–ª–æ –≤ –º–∞—Ç—Ä–∏—Ü–µ

    Examples:
        >>> find_kth_number(3, 5)
        5
    """
    left, right = 1, n * n
    while left < right:
        mid = (left + right) // 2
        count = count_less_equal(n, mid)
        if count < k:
            left = mid + 1
        else:
            right = mid
    return left

# –û—Å–Ω–æ–≤–Ω–∞—è —á–∞—Å—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—ã
if __name__ == "__main__":
    import sys
    input = sys.stdin.read
    data = input().split()
    n = int(data[0])
    k = int(data[1])
    print(find_kth_number(n, k))
````

### –í–ê–®–ê –∏–Ω—Ç—É–∏—Ü–∏—è

–ï—Å–ª–∏ –º—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ø—Ä—è–º–æ–π –ø—Ä–æ—Ö–æ–¥, —Ç–æ $n$ –±–æ–ª—å—à–æ–π (–Ω–∞–ø—Ä–∏–º–µ—Ä, 100‚ÄØ000), —Ç–æ —Ç–∞–±–ª–∏—Ü–∞ –±—É–¥–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å $10^{10}$ —ç–ª–µ–º–µ–Ω—Ç–æ–≤. –°–æ–∑–¥–∞—Ç—å –µ—ë –∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –≤—Å–µ —á–∏—Å–ª–∞ –∑–∞–π–º–µ—Ç —Å–ª–∏—à–∫–æ–º –º–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ –∏ –ø–∞–º—è—Ç–∏. –ü–æ—ç—Ç–æ–º—É –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ —Å–ø–æ—Å–æ–± —Ä–µ—à–∏—Ç—å –∑–∞–¥–∞—á—É –±–µ–∑ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≤—Å–µ–π —Ç–∞–±–ª–∏—Ü—ã.  

–ú—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –±–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫ –ø–æ –∑–Ω–∞—á–µ–Ω–∏—è–º –æ—Ç $1$ –¥–æ $n^2$. –î–ª—è –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–∞ $mid$ –≤ –±–∏–Ω–∞—Ä–Ω–æ–º –ø–æ–∏—Å–∫–µ –º—ã —Å—á–∏—Ç–∞–µ–º, —Å–∫–æ–ª—å–∫–æ —á–∏—Å–µ–ª –≤ —Ç–∞–±–ª–∏—Ü–µ –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ $mid$. –≠—Ç–æ –¥–µ–ª–∞–µ—Ç—Å—è –±–µ–∑ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è —Ç–∞–±–ª–∏—Ü—ã: –¥–ª—è –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏ $i$ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª, –º–µ–Ω—å—à–∏—Ö –∏–ª–∏ —Ä–∞–≤–Ω—ã—Ö $mid$, —Ä–∞–≤–Ω–æ $\min(mid // i, n)$. –°—É–º–º–∏—Ä—É—è —ç—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –≤—Å–µ—Ö —Å—Ç—Ä–æ–∫, –º—ã –ø–æ–ª—É—á–∞–µ–º –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª, –º–µ–Ω—å—à–∏—Ö –∏–ª–∏ —Ä–∞–≤–Ω—ã—Ö $mid$.  

–ï—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª, –º–µ–Ω—å—à–∏—Ö –∏–ª–∏ —Ä–∞–≤–Ω—ã—Ö $mid$, –º–µ–Ω—å—à–µ $k$, –∑–Ω–∞—á–∏—Ç, –∏—Å–∫–æ–º–æ–µ —á–∏—Å–ª–æ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –ø—Ä–∞–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω–µ –¥–∏–∞–ø–∞–∑–æ–Ω–∞. –ò–Ω–∞—á–µ ‚Äî –≤ –ª–µ–≤–æ–π. –ü–æ–≤—Ç–æ—Ä—è—è —ç—Ç–æ—Ç –ø—Ä–æ—Ü–µ—Å—Å, –º—ã —Å—É–∂–∞–µ–º –¥–∏–∞–ø–∞–∑–æ–Ω –¥–æ —Ç–µ—Ö –ø–æ—Ä, –ø–æ–∫–∞ –Ω–µ –Ω–∞–π–¥–µ–º —Ç–æ—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ $k$-–≥–æ –Ω–∞–∏–º–µ–Ω—å—à–µ–≥–æ —á–∏—Å–ª–∞. –≠—Ç–æ—Ç –ø–æ–¥—Ö–æ–¥ —Ä–∞–±–æ—Ç–∞–µ—Ç –∑–∞ $O(n \log n)$, —á—Ç–æ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ –¥–∞–∂–µ –¥–ª—è –±–æ–ª—å—à–∏—Ö $n$.

## K-—è —Å—É–º–º–∞ ([—Å—Å—ã–ª–∫–∞](https://codeforces.com/edu/course/2/lesson/6/5/practice/contest/285084/problem/C)) [1/3 –±–∞–ª–ª–∞]

–ï—Å—Ç—å –¥–≤–∞ –º–∞—Å—Å–∏–≤–∞ ùëé –∏ ùëè, –∫–∞–∂–¥—ã–π –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ ùëõ —á–∏—Å–µ–ª. –î–ª—è –∫–∞–∂–¥–æ–π –ø–∞—Ä—ã —á–∏—Å–µ–ª (ùëñ,ùëó): 1 ‚â§ ùëñ,ùëó ‚â§ ùëõ –≤—ã–ø–∏—à–µ–º —Å—É–º–º—É —á–∏—Å–µ–ª ùëéùëñ + ùëèùëó. –ù–∞–π–¥–∏—Ç–µ –≤ –ø–æ–ª—É—á–µ–Ω–Ω–æ–º –º–Ω–æ–∂–µ—Å—Ç–≤–µ —Å—É–º–º ùëò-—é –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é.

### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
–ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ ùëõ –∏ ùëò (1 ‚â§ ùëõ ‚â§ 10^5, 1 ‚â§ ùëò ‚â§ ùëõ^2). –í—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Å—Å–∏–≤–∞ ùëé, —Ç—Ä–µ—Ç—å—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Å—Å–∏–≤–∞ ùëè. –í—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Å—Å–∏–≤–æ–≤ ‚Äî —Ü–µ–ª—ã–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ —á–∏—Å–ª–∞, –Ω–µ –±–æ–ª—å—à–µ 10^9.

### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
–í—ã–≤–µ–¥–∏—Ç–µ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äî –∏—Å–∫–æ–º—É—é ùëò-—é —Å—É–º–º—É.

### –ü—Ä–∏–º–µ—Ä  

#### –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
5 10
4 2 6 4 8
7 3 1 9 5
```  

#### –í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ  
```
9
```  

````python
from typing import List

def count_less_equal(a: List[int], b: List[int], mid: int) -> int:
    """
    Description:
        –ü–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä (ai, bj), —Ç–∞–∫–∏—Ö —á—Ç–æ ai + bj <= mid.

    Args:
        a: –ü–µ—Ä–≤—ã–π –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –º–∞—Å—Å–∏–≤
        b: –í—Ç–æ—Ä–æ–π –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –º–∞—Å—Å–∏–≤
        mid: –ó–Ω–∞—á–µ–Ω–∏–µ, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ –ø–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä

    Returns:
        –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä, —Å—É–º–º–∞ –∫–æ—Ç–æ—Ä—ã—Ö –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–∞ mid

    Examples:
        >>> count_less_equal([1, 2, 3], [4, 5, 6], 5)
        3
    """
    count = 0
    j = len(b) - 1  # –£–∫–∞–∑–∞—Ç–µ–ª—å –¥–ª—è –º–∞—Å—Å–∏–≤–∞ b
    for ai in a:
        while j >= 0 and ai + b[j] > mid:
            j -= 1
        count += j + 1
    return count

def find_kth_sum(n: int, k: int, a: List[int], b: List[int]) -> int:
    """
    Description:
        –ù–∞—Ö–æ–¥–∏—Ç k-—é –ø–æ –≤–µ–ª–∏—á–∏–Ω–µ —Å—É–º–º—É –ø–∞—Ä —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∏–∑ –º–∞—Å—Å–∏–≤–æ–≤ a –∏ b.

    Args:
        n: –†–∞–∑–º–µ—Ä –º–∞—Å—Å–∏–≤–æ–≤ a –∏ b
        k: –ü–æ—Ä—è–¥–∫–æ–≤—ã–π –Ω–æ–º–µ—Ä —Å—É–º–º—ã
        a: –ü–µ—Ä–≤—ã–π –º–∞—Å—Å–∏–≤
        b: –í—Ç–æ—Ä–æ–π –º–∞—Å—Å–∏–≤

    Returns:
        k-—è –ø–æ –≤–µ–ª–∏—á–∏–Ω–µ —Å—É–º–º–∞ –ø–∞—Ä —ç–ª–µ–º–µ–Ω—Ç–æ–≤

    Examples:
        >>> find_kth_sum(3, 5, [1, 2, 3], [4, 5, 6])
        7
    """
    a.sort()
    b.sort()
    left, right = a[0] + b[0], a[-1] + b[-1]
    while left < right:
        mid = (left + right) // 2
        count = count_less_equal(a, b, mid)
        if count < k:
            left = mid + 1
        else:
            right = mid
    return left

# –ß—Ç–µ–Ω–∏–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
n, k = map(int, input().split())
a = list(map(int, input().split()))
b = list(map(int, input().split()))

# –ü–æ–∏—Å–∫ k-–π —Å—É–º–º—ã
print(find_kth_sum(n, k, a, b))
````

–ò—Å–ø–æ–ª—å–∑—É–µ–º –±–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫. –ú—ã –∑–Ω–∞–µ–º, —á—Ç–æ –º–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Å—É–º–º–∞ ‚Äî —ç—Ç–æ $a[0] + b[0]$, –∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è ‚Äî $a[-1] + b[-1]$. –ë—É–¥–µ–º "—É–≥–∞–¥—ã–≤–∞—Ç—å" $k$-—é —Å—É–º–º—É —Å –ø–æ–º–æ—â—å—é –±–∏–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞ –≤ —ç—Ç–æ–º –¥–∏–∞–ø–∞–∑–æ–Ω–µ.

–î–ª—è –∫–∞–∂–¥–æ–≥–æ "—É–≥–∞–¥–∞–Ω–Ω–æ–≥–æ" —á–∏—Å–ª–∞ $mid$ –º—ã —Å—á–∏—Ç–∞–µ–º, —Å–∫–æ–ª—å–∫–æ —Å—É–º–º $a_i + b_j$ –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–æ $mid$. –ï—Å–ª–∏ —Ç–∞–∫–∏—Ö —Å—É–º–º –º–µ–Ω—å—à–µ $k$, –∑–Ω–∞—á–∏—Ç, $k$-—è —Å—É–º–º–∞ –±–æ–ª—å—à–µ $mid$, –∏ –º—ã –∏—â–µ–º –≤ –ø—Ä–∞–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω–µ. –ò–Ω–∞—á–µ ‚Äî –≤ –ª–µ–≤–æ–π.


–ß—Ç–æ–±—ã –Ω–µ –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å –≤—Å–µ –ø–∞—Ä—ã, –º—ã —Å–æ—Ä—Ç–∏—Ä—É–µ–º –æ–±–∞ —Å–ø–∏—Å–∫–∞. –ó–∞—Ç–µ–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–∞ –∏–∑ —Å–ø–∏—Å–∫–∞ $a$ —Å–º–æ—Ç—Ä–∏–º, —Å–∫–æ–ª—å–∫–æ —á–∏—Å–µ–ª –∏–∑ —Å–ø–∏—Å–∫–∞ $b$ –º–æ–∂–Ω–æ –¥–æ–±–∞–≤–∏—Ç—å –∫ –Ω–µ–º—É, —á—Ç–æ–±—ã —Å—É–º–º–∞ –±—ã–ª–∞ –º–µ–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–∞ $mid$. –≠—Ç–æ –¥–µ–ª–∞–µ—Ç—Å—è —Å –ø–æ–º–æ—â—å—é –¥–≤—É—Ö —É–∫–∞–∑–∞—Ç–µ–ª–µ–π, —á—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç –æ—á–µ–Ω—å –±—ã—Å—Ç—Ä–æ.
