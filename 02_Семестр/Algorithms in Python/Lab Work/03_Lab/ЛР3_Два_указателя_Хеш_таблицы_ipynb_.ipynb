## Задача 217. Contains Duplicate ([ссылка](https://leetcode.com/problems/contains-duplicate/description/)) [0.75 балла]

Перевод задачи:

Дан массив целых чисел `nums`.

Необходимо определить, встречается ли в массиве какое-либо значение как минимум дважды. Если такое значение существует, вернуть `true`; если все элементы различны, вернуть `false`.

---

## Примеры

### Пример 1:
**Вход:**
```python
nums = [1,2,3,1]
```
**Выход:**
```
true
```
**Объяснение:**
Элемент `1` встречается на позициях 0 и 3.

### Пример 2:
**Вход:**
```python
nums = [1,2,3,4]
```
**Выход:**
```
false
```
**Объяснение:**
Все элементы в массиве уникальны.

### Пример 3:
**Вход:**
```python
nums = [1,1,1,3,3,4,3,2,4,2]
```
**Выход:**
```
true
```

---

## Ограничения:
- $1 \leq nums.length \leq 10^5$
- $-10^9 \leq nums[i] \leq 10^9$

````python
# Асимптотическая сложность по времени - О(TODO)
# Асимптотическая сложность по памяти - O(TODO)

# ВАШ КОД
from typing import List

class Solution:
    """
    Description:
    ---------------
        Класс для решения задачи поиска дубликатов в списке.
    """

    def containsDuplicate(self, nums: List[int]) -> bool:
        """
        Description:
        ---------------
            Проверяет, содержит ли список дубликаты.

        Args:
        ---------------
            nums: Список целых чисел.

        Returns:
        ---------------
            True, если в списке есть дубликаты, иначе False.

        Examples:
        ---------------
            >>> sol = Solution()
            >>> sol.containsDuplicate([1, 2, 3, 1])
            True
            >>> sol.containsDuplicate([1, 2, 3, 4])
            False
        """
        # Создаем множество для отслеживания уже виденных чисел
        seen = set()

        # Проходим по каждому числу в списке
        for num in nums:
            # Если число уже встречалось, возвращаем True
            if num in seen:
                return True
            # Добавляем число в множество виденных чисел
            seen.add(num)

        # Если дубликатов не найдено, возвращаем False
        return False
````

### ВАША интуиция

Для этой задачи идеально подходит хеш-таблица (множество в Python), которая обеспечивает константное время доступа. Таким образом, мы можем просматривать элементы один за другим и проверять, встречался ли текущий элемент ранее.

При обходе массива каждый элемент проверяется на наличие в множестве. Если элемент уже присутствует, значит он встречается как минимум дважды, и мы сразу возвращаем True. Если элемент отсутствует, добавляем его в множество. Если после просмотра всех элементов дубликаты не обнаружены, возвращаем False.

В худшем случае, когда дубликатов нет, нам придётся просмотреть весь массив и сохранить все его элементы в множестве, что даёт линейную сложность как по времени, так и по памяти. Однако, эта сложность оптимальна, поскольку невозможно решить задачу без просмотра каждого элемента хотя бы один раз.

## Задача 219. Contains Duplicate II ([ссылка](https://leetcode.com/problems/contains-duplicate-ii/description/)) [0.75 балла]

Перевод задачи:
Дан массив целых чисел `nums` и целое число `k`.

Необходимо проверить, существуют ли два различных индекса `i` и `j` в массиве, для которых выполняются условия:
- `nums[i] == nums[j]`
- $|i - j| \leq k$

Если такие индексы существуют, вернуть `true`, иначе вернуть `false`.

---

## Примеры

### Пример 1:
**Вход:**
```python
nums = [1,2,3,1]
k = 3
```
**Выход:**
```
true
```
**Объяснение:**
Элемент `1` встречается на позициях 0 и 3, расстояние между ними равно 3, что удовлетворяет условию $|0-3| \leq 3$.

### Пример 2:
**Вход:**
```python
nums = [1,0,1,1]
k = 1
```
**Выход:**
```
true
```
**Объяснение:**
Элемент `1` встречается на позициях 2 и 3, расстояние между ними равно 1, что удовлетворяет условию $|2-3| \leq 1$.

### Пример 3:
**Вход:**
```python
nums = [1,2,3,1,2,3]
k = 2
```
**Выход:**
```
false
```
**Объяснение:**
Хотя элементы повторяются, ни одна пара равных элементов не удовлетворяет условию $|i - j| \leq 2$.

---

## Ограничения:
- $1 \leq nums.length \leq 10^5$
- $-10^9 \leq nums[i] \leq 10^9$
- $0 \leq k \leq 10^5$

````python
# Асимптотическая сложность по времени - О(TODO)
# Асимптотическая сложность по памяти - O(TODO)

# ВАШ КОД
from typing import List

class Solution:
    """
    Description:
    ---------------
        Класс для решения задачи поиска близких дубликатов в списке.
    """

    def containsNearbyDuplicate(self, nums: List[int], k: int) -> bool:
        """
        Description:
        ---------------
            Проверяет, содержит ли список дубликаты, расстояние между которыми не превышает k.

        Args:
        ---------------
            nums: Список целых чисел.
            k: Максимальное допустимое расстояние между дубликатами.

        Returns:
        ---------------
            True, если в списке есть близкие дубликаты, иначе False.

        Examples:
        ---------------
            >>> sol = Solution()
            >>> sol.containsNearbyDuplicate([1, 2, 3, 1], 3)
            True
            >>> sol.containsNearbyDuplicate([1, 0, 1, 1], 1)
            True
            >>> sol.containsNearbyDuplicate([1, 2, 3, 4], 2)
            False
        """
        # Создаем хеш-таблицу для хранения последнего индекса каждого элемента
        index_map = {}

        # Проходим по каждому элементу списка с его индексом
        for i, num in enumerate(nums):
            # Если элемент уже встречался и расстояние не превышает k
            if num in index_map and i - index_map[num] <= k:
                return True
            # Обновляем или добавляем текущий индекс элемента
            index_map[num] = i

        # Если близких дубликатов не найдено, возвращаем False
        return False
````

### ВАША интуиция

Для эффективного решения мы используем хеш-таблицу (словарь в Python), где ключом является значение элемента массива, а значением - индекс его последнего вхождения. При итерации по массиву, для каждого элемента мы проверяем, встречался ли он ранее, и если да, то не превышает ли разница между текущим и предыдущим индексом значение k. Если оба условия выполняются, то мы нашли пару индексов, удовлетворяющую условиям задачи.

Этот алгоритм эффективен, поскольку он требует всего одного прохода по массиву и использует хеш-таблицу для быстрого поиска предыдущих вхождений элементов. Хотя в худшем случае мы используем O(n) дополнительной памяти, это необходимо для хранения информации о предыдущих индексах элементов, и решение остается оптимальным с точки зрения асимптотической сложности.

## Задача 242. Valid Anagram ([ссылка](https://leetcode.com/problems/valid-anagram/description/)) [0.75 балла]

Перевод задачи:
Даны две строки `s` и `t`.

Необходимо определить, является ли строка `t` анаграммой строки `s`. Другими словами, проверить, можно ли перестановкой символов строки `s` получить строку `t`. Если это так, вернуть `true`, иначе — `false`.

---

## Примеры

### Пример 1:
**Вход:**
```python
s = "anagram"
t = "nagaram"
```
**Выход:**
```
true
```

### Пример 2:
**Вход:**
```python
s = "rat"
t = "car"
```
**Выход:**
```
false
```

---

## Ограничения:
- $1 \leq s.length, t.length \leq 5 \times 10^4$
- Строки `s` и `t` состоят из строчных латинских букв.

---

## Дополнительный вопрос:
**Follow up:** Что если входные данные содержат символы Юникода?  

````python
# Асимптотическая сложность по времени - О(TODO)
# Асимптотическая сложность по памяти - O(TODO)

# ВАШ КОД
from typing import Dict

class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        """
        Description:
        ---------------
            Проверяет, являются ли две строки анаграммами.

        Args:
        ---------------
            s: Первая строка для проверки
            t: Вторая строка для проверки

        Returns:
        ---------------
            True, если строки являются анаграммами, иначе False

        Raises:
        ---------------
            Нет исключений

        Examples:
        ---------------
            >>> sol = Solution()
            >>> sol.isAnagram("listen", "silent")
            True
            >>> sol.isAnagram("hello", "world")
            False
        """
        # Если длины строк не равны, они не могут быть анаграммами
        if len(s) != len(t):
            return False

        # Словарь для подсчета символов в строке s
        char_count: Dict[str, int] = {}

        # Подсчет символов в строке s
        for char in s:
            char_count[char] = char_count.get(char, 0) + 1

        # Проверка символов в строке t
        for char in t:
            # Если символ отсутствует в словаре или его количество равно нулю,
            # строки не являются анаграммами
            if char not in char_count or char_count[char] == 0:
                return False
            char_count[char] -= 1

        return True
````

### ВАША интуиция

Ключевая идея этого решения заключается в том, что две строки являются анаграммами тогда и только тогда, когда они содержат одинаковое количество каждого уникального символа. Проще говоря, частотный анализ символов в обеих строках должен совпадать.

Алгоритм реализует этот подход в два этапа: сначала подсчитывает частоту каждого символа в первой строке, затем проверяет вторую строку, уменьшая соответствующие счетчики. Если в какой-то момент встречается символ, отсутствующий в первой строке или превышающий свою частоту, алгоритм немедленно возвращает false. Важно отметить быструю проверку на равенство длин строк в начале, так как строки разной длины не могут быть анаграммами.

Что касается работы с Unicode (из дополнительного вопроса), решение с использованием словаря уже поддерживает это, так как в Python строки представлены в Unicode, и словарь может хранить любые символы в качестве ключей. Для других языков программирования может потребоваться специальная обработка, учитывающая многобайтовую природу символов Unicode, но принцип решения остается тем же.

## Задача 205. Isomorphic Strings ([ссылка](https://leetcode.com/problems/isomorphic-strings/description/)) [0.75 балла]

Перевод задачи:
Даны две строки `s` и `t`. Строки называются изоморфными, если символы строки `s` можно заменить так, чтобы получилась строка `t`.

При этом:
- Все вхождения одного и того же символа в строке `s` должны быть заменены на один и тот же символ.
- При замене сохраняется порядок символов.
- Никакие два разных символа не могут быть заменены на один и тот же символ, однако символ может быть заменен сам на себя.

---

## Примеры

### Пример 1:
**Вход:**
```python
s = "egg"
t = "add"
```
**Выход:**
```
true
```
**Объяснение:**
Строки можно сделать идентичными, если:
- Заменить `'e'` на `'a'`.
- Заменить `'g'` на `'d'`.

### Пример 2:
**Вход:**
```python
s = "foo"
t = "bar"
```
**Выход:**
```
false
```
**Объяснение:**
Невозможно сделать строки идентичными, так как символ `'o'` должен быть заменён одновременно на `'a'` и `'r'`.

### Пример 3:
**Вход:**
```python
s = "paper"
t = "title"
```
**Выход:**
```
true
```

---

## Ограничения:
- $1 \leq s.length \leq 5 \times 10^4$
- Длина строки `t` равна длине строки `s`
- Строки `s` и `t` состоят из любых допустимых символов ASCII

````python
# Асимптотическая сложность по времени - О(TODO)
# Асимптотическая сложность по памяти - O(TODO)

# ВАШ КОД
from typing import Dict

class Solution:
    def isIsomorphic(self, s: str, t: str) -> bool:
        """
        Description:
        ---------------
            Проверяет, являются ли две строки изоморфными.
            Две строки s и t изоморфны, если символы в s можно
            заменить на символы в t, сохраняя порядок.

        Args:
        ---------------
            s: Первая строка для проверки.
            t: Вторая строка для проверки.

        Returns:
        ---------------
            bool: True, если строки изоморфны, иначе False.

        Raises:
        ---------------
            Нет исключений.

        Examples:
        ---------------
            >>> solution = Solution()
            >>> solution.isIsomorphic("egg", "add")
            True
            >>> solution.isIsomorphic("foo", "bar")
            False
        """
        # Словари для отображения символов
        mapping_s_to_t: Dict[str, str] = {}
        mapping_t_to_s: Dict[str, str] = {}

        # Проходим по парам символов из обеих строк
        for char_s, char_t in zip(s, t):
            # Проверяем, соответствует ли существующее отображение s -> t
            if char_s in mapping_s_to_t and mapping_s_to_t[char_s] != char_t:
                return False

            # Проверяем, соответствует ли существующее отображение t -> s
            if char_t in mapping_t_to_s and mapping_t_to_s[char_t] != char_s:
                return False

            # Добавляем новые отображения
            mapping_s_to_t[char_s] = char_t
            mapping_t_to_s[char_t] = char_s

        return True
````

### ВАША интуиция

Представьте, что вы переводите с одного языка на другой, где каждый символ заменяется строго определённым другим символом. Когда мы проверяем изоморфизм строк, мы фактически создаём "словарь перевода", где каждой букве из первой строки соответствует ровно одна буква из второй строки.

Ключевой момент в этой задаче — обеспечить взаимно-однозначное соответствие. Это означает, что не только каждый символ первой строки должен всегда переводиться в один и тот же символ второй строки, но и каждый символ второй строки должен иметь только один "источник" в первой строке. Для проверки этого мы используем два словаря: один хранит соответствие от s к t, а другой — от t к s.

Алгоритм последовательно сравнивает символы обеих строк и проверяет согласованность отображений. Если нам встречается новая пара символов, мы добавляем её в оба словаря. Если же мы видим символ, который уже есть в словаре, но его перевод не соответствует текущему символу второй строки, значит строки не изоморфны. Аналогично, если символ второй строки уже имеет другой источник в первой строке, изоморфизм невозможен.

## Задача 2615. Sum of Distances ([ссылка](https://leetcode.com/problems/sum-of-distances/description/)) [0.5 балла]

Перевод задачи:
Дан массив целых чисел `nums` с индексированием с нуля.

Существует массив `arr` длины `nums.length`, где для каждого индекса `i` выполняется следующее условие:
- `arr[i]` равен сумме значений `|i - j|` для всех индексов `j`, таких что `nums[j] == nums[i]` и `j != i`.
- Если таких индексов `j` не существует, то `arr[i]` устанавливается равным 0.

Необходимо вернуть массив `arr`.

---

## Примеры

### Пример 1:
**Вход:**
```python
nums = [1, 3, 1, 1, 2]
```
**Выход:**
```
[5, 0, 3, 4, 0]
```
**Объяснение:**
- Для `i = 0`:  
  Элемент `nums[0]` равен `1`, совпадает с `nums[2]` и `nums[3]`.  
  Таким образом, `arr[0] = |0 - 2| + |0 - 3| = 2 + 3 = 5`.
- Для `i = 1`:  
  Элемент `nums[1]` равен `3`, других элементов со значением `3` нет, поэтому `arr[1] = 0`.
- Для `i = 2`:  
  Элемент `nums[2]` равен `1`, совпадает с `nums[0]` и `nums[3]`.  
  Таким образом, `arr[2] = |2 - 0| + |2 - 3| = 2 + 1 = 3`.
- Для `i = 3`:  
  Элемент `nums[3]` равен `1`, совпадает с `nums[0]` и `nums[2]`.  
  Таким образом, `arr[3] = |3 - 0| + |3 - 2| = 3 + 1 = 4`.
- Для `i = 4`:  
  Элемент `nums[4]` равен `2`, других элементов со значением `2` нет, поэтому `arr[4] = 0`.

### Пример 2:
**Вход:**
```python
nums = [0, 5, 3]
```
**Выход:**
```
[0, 0, 0]
```
**Объяснение:**
Все элементы в массиве различны, поэтому для каждого индекса `i` других индексов с таким же значением не существует, и `arr[i] = 0`.

---

## Ограничения:
- $1 \leq nums.length \leq 10^5$
- $0 \leq nums[i] \leq 10^9$

````python
# Асимптотическая сложность по времени - О(TODO)
# Асимптотическая сложность по памяти - O(TODO)

# ВАШ КОД
from typing import List, Dict

class Solution:
    def distance(self, nums: List[int]) -> List[int]:
        """
        Description:
        ---------------
            Вычисляет сумму расстояний для каждого элемента в списке
            до всех остальных элементов с таким же значением.

        Args:
        ---------------
            nums: Список целых чисел.

        Returns:
        ---------------
            List[int]: Список, где каждый элемент представляет собой
                       сумму расстояний до всех остальных элементов
                       с таким же значением.

        Raises:
        ---------------
            Нет исключений.

        Examples:
        ---------------
            >>> solution = Solution()
            >>> solution.distance([1, 3, 1, 1, 2])
            [5, 0, 3, 2, 0]
        """
        n = len(nums)
        result = [0] * n

        # Словарь для группировки индексов по значениям
        value_to_indices: Dict[int, List[int]] = {}

        # Заполняем словарь
        for i, num in enumerate(nums):
            if num not in value_to_indices:
                value_to_indices[num] = []
            value_to_indices[num].append(i)

        # Обрабатываем каждую группу
        for num, indices in value_to_indices.items():
            # Если элемент уникальный, пропускаем
            if len(indices) <= 1:
                continue

            # Вычисляем префиксные суммы для каждого индекса
            prefix_sum = [0] * (len(indices) + 1)
            for i in range(len(indices)):
                prefix_sum[i + 1] = prefix_sum[i] + indices[i]

            # Вычисляем сумму расстояний для каждого индекса
            for i, idx in enumerate(indices):
                # Сумма расстояний до элементов слева
                left_sum = idx * i - prefix_sum[i]

                # Сумма расстояний до элементов справа
                right_sum = (prefix_sum[len(indices)] - prefix_sum[i + 1]) - (len(indices) - i - 1) * idx

                # Общая сумма
                result[idx] = left_sum + right_sum

        return result
````

### ВАША интуиция

Ключевая идея решения заключается в группировке индексов по их значениям, что позволяет избежать повторных вычислений. Вместо того чтобы перебирать весь массив для каждого элемента (что привело бы к квадратичной сложности), мы собираем элементы с одинаковыми значениями в группы и работаем с каждой группой отдельно.

## Задача 2799. Count Complete Subarrays in an Array ([ссылка](https://leetcode.com/problems/count-complete-subarrays-in-an-array/description/)) [0.5 балла]

Перевод задачи:
Дан массив `nums`, состоящий из положительных целых чисел.

Подмассив массива называется полным, если выполняется следующее условие:
- Количество различных элементов в подмассиве равно количеству различных элементов во всём массиве `nums`.

Необходимо вернуть количество полных подмассивов.

**Примечание:** Подмассив – это непустая, смежная последовательность элементов исходного массива.

---

## Примеры

### Пример 1:
**Вход:**
```python
nums = [1,3,1,2,2]
```
**Выход:**
```
4
```
**Объяснение:**
Полными подмассивами являются следующие:
- `[1,3,1,2]`
- `[1,3,1,2,2]`
- `[3,1,2]`
- `[3,1,2,2]`

### Пример 2:
**Вход:**
```python
nums = [5,5,5,5]
```
**Выход:**
```
10
```
**Объяснение:**
Так как в массиве содержится только число `5`, любой подмассив будет полным. Общее число возможных подмассивов равно 10.

---

## Ограничения:
- $1 \leq nums.length \leq 1000$
- $1 \leq nums[i] \leq 2000$


````python
# Асимптотическая сложность по времени - О(TODO)
# Асимптотическая сложность по памяти - O(TODO)

# ВАШ КОД
from typing import List

class Solution:
    def countCompleteSubarrays(self, nums: List[int]) -> int:
        """
        Description:
        ---------------
            Подсчитывает количество полных подмассивов в списке nums.
            Полный подмассив - это подмассив, содержащий все уникальные
            элементы из nums.

        Args:
        ---------------
            nums: Список целых чисел

        Returns:
        ---------------
            Количество полных подмассивов

        Raises:
        ---------------
            None

        Examples:
        ---------------
            >>> sol = Solution()
            >>> sol.countCompleteSubarrays([1, 3, 1, 2, 2])
            4
        """
        # Общее количество уникальных элементов
        total_distinct = len(set(nums))
        n = len(nums)

        # Если всего один уникальный элемент, все подмассивы полные
        if total_distinct == 1:
            return n * (n + 1) // 2

        count = 0
        left = 0
        window_freq = {}

        # Для каждой правой границы окна
        for right in range(n):
            # Добавляем элемент в окно
            window_freq[nums[right]] = window_freq.get(nums[right], 0) + 1

            # Пока окно содержит все уникальные элементы
            while len(window_freq) == total_distinct:
                # Уменьшаем частоту левого элемента
                window_freq[nums[left]] -= 1
                if window_freq[nums[left]] == 0:
                    del window_freq[nums[left]]

                # Все подмассивы от текущего left+1 до n-1 с правой границей right
                # будут полными
                count += n - right

                left += 1

        return count
````

### ВАША интуиция

Ключ к решению этой задачи заключается в понимании, что как только подмассив становится "полным" (содержит все уникальные элементы из исходного массива), любое расширение этого подмассива вправо также будет полным.

Я использую метод скользящего окна, постепенно сдвигая левую границу окна и подсчитывая все возможные полные подмассивы. Когда текущее окно содержит все уникальные элементы, я могу подсчитать, сколько полных подмассивов начинается с текущей левой границы, а затем сдвинуть эту границу вправо, чтобы найти новые минимальные полные подмассивы.

## Задача 1577. Number of Ways Where Square of Number Is Equal to Product of Two Numbers ([ссылка](https://leetcode.com/problems/number-of-ways-where-square-of-number-is-equal-to-product-of-two-numbers/description/)) [1 балл]

Даны два массива целых чисел `nums1` и `nums2`.

Нужно посчитать количество тройек (как типа 1, так и типа 2), удовлетворяющих следующим правилам:

**Тип 1:** Тройка `(i, j, k)` считается валидной, если выполняется условие:
- `nums1[i]^2 == nums2[j] * nums2[k]`
- при этом: `0 <= i < nums1.length` и `0 <= j < k < nums2.length`.

**Тип 2:** Тройка `(i, j, k)` считается валидной, если выполняется условие:
- `nums2[i]^2 == nums1[j] * nums1[k]`
- при этом: `0 <= i < nums2.length` и `0 <= j < k < nums1.length`.

Необходимо вернуть суммарное количество валидных тройек обоих типов.

---

## Примеры

### Пример 1:
**Вход:**
```python
nums1 = [7,4]
nums2 = [5,2,8,9]
```
**Выход:**
```
1
```
**Объяснение:**
- Тип 1: тройка `(1, 1, 2)`, так как `nums1[1]^2 = nums2[1] * nums2[2]` (то есть, $4^2 = 2 \times 8$, $16 = 16$).

### Пример 2:
**Вход:**
```python
nums1 = [1,1]
nums2 = [1,1,1]
```
**Выход:**
```
9
```
**Объяснение:**
Все тройки валидны, поскольку $1^2 = 1 \times 1$.

- Тип 1: тройки `(0,0,1)`, `(0,0,2)`, `(0,1,2)`, `(1,0,1)`, `(1,0,2)`, `(1,1,2)`.
- Тип 2: тройки `(0,0,1)`, `(1,0,1)`, `(2,0,1)`.

### Пример 3:
**Вход:**
```python
nums1 = [7,7,8,3]
nums2 = [1,2,9,7]
```
**Выход:**
```
2
```
**Объяснение:**
- Тип 1: тройка `(3,0,2)`, так как `nums1[3]^2 = nums2[0] * nums2[2]`.
- Тип 2: тройка `(3,0,1)`, так как `nums2[3]^2 = nums1[0] * nums1[1]`.

---

## Ограничения:
- $1 \leq nums1.length, nums2.length \leq 1000$
- $1 \leq nums1[i], \, nums2[i] \leq 10^5$

````python
# Асимптотическая сложность по времени - О(TODO)
# Асимптотическая сложность по памяти - O(TODO)

# ВАШ КОД
from collections import Counter
from typing import List

class Solution:
    def numTriplets(self, nums1: List[int], nums2: List[int]) -> int:
        """
        Description:
        ---------------
            Подсчитывает количество троек (i, j, k), где nums1[i]^2 == nums2[j] * nums2[k].

        Args:
        ---------------
            nums1: Первый список целых чисел
            nums2: Второй список целых чисел

        Returns:
        ---------------
            Количество троек

        Raises:
        ---------------
            None

        Examples:
        ---------------
            >>> sol = Solution()
            >>> sol.numTriplets([7, 4], [5, 2, 8, 9])
            1
        """
        # Вызываем функцию подсчета для обоих типов троек
        return self.count_triplets(nums1, nums2) + self.count_triplets(nums2, nums1)

    def count_triplets(self, nums1: List[int], nums2: List[int]) -> int:
        """
        Description:
        ---------------
            Подсчитывает количество троек (i, j, k), где nums1[i]^2 == nums2[j] * nums2[k].

        Args:
        ---------------
            nums1: Первый список целых чисел
            nums2: Второй список целых чисел

        Returns:
        ---------------
            Количество троек

        Raises:
        ---------------
            None

        Examples:
        ---------------
            >>> sol = Solution()
            >>> sol.count_triplets([7, 4], [5, 2, 8, 9])
            1
        """
        count = 0
        nums2_freq = Counter(nums2)

        # Для каждого элемента в первом массиве
        for num1 in nums1:
            square = num1 * num1

            # Проверяем все пары во втором массиве
            for num2, freq2 in nums2_freq.items():
                # Проверяем, делится ли квадрат на текущий элемент без остатка
                if square % num2 == 0:
                    # Находим второй множитель
                    other = square // num2

                    # Если оба числа одинаковые
                    if other == num2:
                        # Считаем количество пар из одинаковых чисел: C(freq2, 2)
                        count += freq2 * (freq2 - 1) // 2
                    # Если числа разные и other есть в массиве
                    elif other in nums2_freq and other > num2:
                        # Добавляем произведение частот
                        count += freq2 * nums2_freq[other]

        return count
````

### ВАША интуиция

Ключевая идея решения заключается в том, чтобы для каждого числа из первого массива вычислить его квадрат и затем искать пары чисел во втором массиве, произведение которых равно этому квадрату. Вместо того чтобы перебирать все возможные пары (что имеет квадратичную сложность), мы используем словарь для подсчета частоты каждого числа во втором массиве.
